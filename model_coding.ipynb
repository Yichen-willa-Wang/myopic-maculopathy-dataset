### based line : logistic

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, roc_curve, roc_auc_score,
    recall_score, confusion_matrix
)
import matplotlib.pyplot as plt

# -------- Train model ----------
model = LogisticRegression(max_iter=1000, solver='liblinear')
model.fit(X_train, y_train)

# -------- Predict probs + labels ----------
def get_metrics(X, y, label):
    prob = model.predict_proba(X)[:, 1]
    pred = model.predict(X)
    acc = accuracy_score(y, pred)
    auc = roc_auc_score(y, prob)
    rec = recall_score(y, pred)
    fpr, tpr, _ = roc_curve(y, prob)
    print(f"{label:<10} | ACC = {acc:.3f} | AUC = {auc:.3f} | Recall₁ = {rec:.3f}")
    return fpr, tpr, auc

fpr_tr, tpr_tr, auc_tr = get_metrics(X_train, y_train, "Train")
fpr_va, tpr_va, auc_va = get_metrics(X_valid, y_valid, "Valid")
fpr_te, tpr_te, auc_te = get_metrics(X_test,  y_test,  "Test")

# -------- ROC Curve ----------
plt.figure(figsize=(8,6))
plt.plot(fpr_tr, tpr_tr, label=f'Train AUC = {auc_tr:.2f}', linestyle='-')
plt.plot(fpr_va, tpr_va, label=f'Valid AUC = {auc_va:.2f}', linestyle='--')
plt.plot(fpr_te, tpr_te, label=f'Test AUC = {auc_te:.2f}', linestyle='-.')
plt.plot([0,1], [0,1], 'k:', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves - Logistic Regression')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------------------
# 0. Imports
# ---------------------------------------------------------------------
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, roc_auc_score, roc_curve,
                             confusion_matrix, ConfusionMatrixDisplay)
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

# ---------------------------------------------------------------------
# 1. 70‑20‑10 Stratified split  (Step‑1 & Step‑2 in your earlier note)
# ---------------------------------------------------------------------
from sklearn.model_selection import train_test_split
X = df.drop(columns=['final'])
y = df['final']
y_binary = (y == 'T').astype(int)
X_train, X_temp, y_train, y_temp = train_test_split(X, y_binary, test_size=0.3, random_state=42, stratify=y_binary)
X_valid, X_test, y_valid, y_test = train_test_split(X_temp, y_temp, test_size=1/3, random_state=42, stratify=y_temp)

# ---------------------------------------------------------------------
# 2. Logistic‑regression feature selection (with scaling)
# ---------------------------------------------------------------------
scaler   = StandardScaler()
X_train_z = scaler.fit_transform(X_train)      # scale **train only**
logreg   = LogisticRegression(max_iter=2000, penalty="l2", solver="liblinear")
logreg.fit(X_train_z, y_train)

# Top‑K absolute coefficients
K = 20                                         # tweak as you like
coef_abs = np.abs(logreg.coef_[0])
top_idx  = np.argsort(coef_abs)[-K:]
feat_top = X_train.columns[top_idx]            # names of selected features

# Sub‑set all three splits to those K columns (unscaled, tree‑friendly)
X_train_sel = X_train[feat_top]
X_valid_sel = X_valid[feat_top]
X_test_sel  = X_test[feat_top]

# ---------------------------------------------------------------------
# 3. XGBoost + GridSearchCV (with L1 & L2 regularization)
# ---------------------------------------------------------------------
xgb = XGBClassifier(
    objective="binary:logistic",
    eval_metric="auc",
    random_state=42,
    n_jobs=-1,
    use_label_encoder=False
)

param_grid = {
    "n_estimators":      [100, 200, 300],
    "max_depth":         [3, 5, 7],
    "learning_rate":     [0.05, 0.1, 0.2],
    "subsample":         [0.8, 1.0],
    "colsample_bytree":  [0.8, 1.0],
    "reg_alpha":         [0, 0.1, 1],     # L1
    "reg_lambda":        [1, 5, 10]       # L2
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring="roc_auc",
    cv=cv,
    verbose=2,
    n_jobs=-1
)
grid.fit(X_train_sel, y_train)
best_xgb = grid.best_estimator_
print("🚀 Best CV‑AUC:", grid.best_score_)
print("🚀 Best params:", grid.best_params_)

# ---------------------------------------------------------------------
# 4. Helper for evaluation
# ---------------------------------------------------------------------
def _eval(model, X, y, tag):
    y_prob = model.predict_proba(X)[:, 1]
    y_pred = model.predict(X)
    auc  = roc_auc_score(y, y_prob)
    acc  = accuracy_score(y, y_pred)
    cm   = confusion_matrix(y, y_pred)
    print(f"{tag:<6} | AUC = {auc:.4f} | ACC = {acc:.4f}")
    return {"tag": tag, "auc": auc, "acc": acc, "cm": cm, "prob": y_prob, "true": y}

res_train = _eval(best_xgb, X_train_sel, y_train, "Train")
res_val   = _eval(best_xgb, X_valid_sel, y_valid, "Valid")
res_test  = _eval(best_xgb, X_test_sel,  y_test,  "Test")

# ---------------------------------------------------------------------
# 5. ROC curves (all three splits)
# ---------------------------------------------------------------------
plt.figure(figsize=(6,5))
for res, style in zip([res_train, res_val, res_test],
                      ["-", "--", "-."]):
    fpr, tpr, _ = roc_curve(res["true"], res["prob"])
    plt.plot(fpr, tpr, style, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
plt.plot([0,1], [0,1], "k:", lw=1)
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curves – XGBoost (top‑20 LR features)")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()

# ---------------------------------------------------------------------
# 6. Confusion matrices
# ---------------------------------------------------------------------
for res in [res_train, res_val, res_test]:
    ConfusionMatrixDisplay(confusion_matrix=res["cm"]).plot(cmap="Blues")
    plt.title(f'{res["tag"]} Confusion Matrix')
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------------------
# 7. Summary table
# ---------------------------------------------------------------------
summary = pd.DataFrame([
    {"Dataset": r["tag"], "AUC": r["auc"], "Accuracy": r["acc"]}
    for r in [res_train, res_val, res_test]
]).set_index("Dataset")
print("\n=== Performance Summary ===")
print(summary)
```

```python
##############################################################
#  FULL PIPELINE — HalvingRandomSearchCV + early‑stop + Platt
##############################################################

# ---------------- 0. Imports ----------------
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.metrics import (roc_auc_score, accuracy_score, roc_curve,
                             confusion_matrix, ConfusionMatrixDisplay,
                             balanced_accuracy_score, f1_score)
from sklearn.experimental import enable_halving_search_cv  # noqa
from sklearn.model_selection import HalvingRandomSearchCV
import joblib

import xgboost as xgb

# ---------------- 1. Split 7:2:1 ------------
X = df.drop(columns=['final'])
y = (df['final'] == 'T').astype(int)
X_tr, X_tmp, y_tr, y_tmp = train_test_split(X, y, test_size=0.3,
                                            stratify=y, random_state=42)
X_val, X_te, y_val, y_te = train_test_split(X_tmp, y_tmp, test_size=1/3,
                                            stratify=y_tmp, random_state=42)

# ---------------- 2. logistic top‑24 --------
scaler_lr = StandardScaler().fit(X_tr)
logreg    = LogisticRegression(max_iter=2000, solver="liblinear")
logreg.fit(scaler_lr.transform(X_tr), y_tr)
top_idx   = np.argsort(np.abs(logreg.coef_[0]))[-24:]
TOP24     = X_tr.columns[top_idx]

# -------- 3. feature weighter ---------------
class FeatureWeighter(BaseEstimator, TransformerMixin):
    def __init__(self, cols, alpha=2.0):
        self.cols, self.alpha = cols, alpha
    def fit(self, X, y=None): return self
    def transform(self, X):
        Xw = X.copy()
        Xw[self.cols] *= self.alpha
        return Xw

# -------- 4. Pipeline -----------------------
xgb_base = XGBClassifier(
    objective="binary:logistic",
    eval_metric="auc",
    random_state=42,
    n_jobs=1,                 # 交给外层并行
    tree_method="hist"        # CPU; 改 'hist', device='cuda' 可用 GPU
)

pipe = Pipeline([
    ("weight", FeatureWeighter(cols=TOP24)),
    ("scale",  StandardScaler()),
    ("xgb",    xgb_base)
])

# -------- 5. param_grid ---------------------
pos_ratio = (y_tr == 1).mean()
spw0      = (1-pos_ratio)/pos_ratio
param_grid = {
    "weight__alpha":         [1.5, 2.0],
    "xgb__n_estimators":     [400, 600],       # ←★ 适当缩小
    "xgb__learning_rate":    [0.01],
    "xgb__max_depth":        [2, 3],
    "xgb__min_child_weight": [5],
    "xgb__gamma":            [0, 1],
    "xgb__subsample":        [0.8],
    "xgb__colsample_bytree": [0.8],
    "xgb__reg_alpha":        [1, 5],
    "xgb__reg_lambda":       [10, 50],
    "xgb__scale_pos_weight": [spw0, spw0*1.5]
}

search = HalvingRandomSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    n_candidates=60, factor=3,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring="roc_auc",
    resource="n_samples",
    max_resources="auto",
    verbose=3, n_jobs=-1, random_state=42
)

# ★★★  这里不再传早停参数，只保留静默输出  ★★★
fit_params = {
    "xgb__verbose": False                # 如要看每棵树的 eval，可删掉这行
}

with joblib.parallel_backend("loky", inner_max_num_threads=1):
    search.fit(X_tr, y_tr, **fit_params)

best_alpha = search.best_params_["weight__alpha"]
best_xgb   = {k.split("__")[1]: v for k, v in search.best_params_.items()
              if k.startswith("xgb__")}

# ---------------- 6. Pre‑processing ---------
def weight(df, alpha):
    dfc = df.copy()
    dfc[TOP24] *= alpha
    return dfc

X_tv   = pd.concat([X_tr, X_val]); y_tv = pd.concat([y_tr, y_val])
scaler = StandardScaler().fit(weight(X_tv, best_alpha))
X_tv_s = scaler.transform(weight(X_tv, best_alpha))
X_val_s= scaler.transform(weight(X_val, best_alpha))
X_te_s = scaler.transform(weight(X_te, best_alpha))
X_tr_s = scaler.transform(weight(X_tr, best_alpha))

# ---------------- 7. bagging (5×) ----------
prob_val = np.zeros(len(X_val))
prob_te  = np.zeros(len(X_te))
for seed in range(5):
    boot   = np.random.choice(len(X_tv_s), len(X_tv_s), replace=True)
    model  = XGBClassifier(**best_xgb, random_state=seed,
                           objective="binary:logistic",
                           eval_metric="auc",
                           n_jobs=1, tree_method="hist")
    model.fit(X_tv_s[boot], y_tv.values[boot])
    prob_val += model.predict_proba(X_val_s)[:,1]
    prob_te  += model.predict_proba(X_te_s)[:,1]
prob_val /= 5; prob_te /= 5

# ---------------- 8. Platt scaling ----------
platt = LogisticRegression(solver="lbfgs").fit(
    prob_val.reshape(-1,1), y_val
)
prob_val_cal = platt.predict_proba(prob_val.reshape(-1,1))[:,1]
prob_te_cal  = platt.predict_proba(prob_te.reshape(-1,1))[:,1]

# choose threshold that maximises balanced accuracy on valid
ths = np.linspace(0.1, 0.9, 81)
bal_scores = [balanced_accuracy_score(y_val, (prob_val_cal>=t)) for t in ths]
best_thr   = ths[int(np.argmax(bal_scores))]
print(f"\n📐 Threshold (max balanced‑ACC) = {best_thr:.2f}")

# ---------------- 9. evaluation ------------
def evaluate(y, prob, thr, tag):
    pred = (prob >= thr).astype(int)
    auc  = roc_auc_score(y, prob)
    acc  = accuracy_score(y, pred)
    cm   = confusion_matrix(y, pred)
    rec  = cm[1,1] / (cm[1,1]+cm[1,0])
    print(f"{tag:<6}| AUC={auc:.3f} | ACC={acc:.3f} | Recall₁={rec:.3f}")
    return dict(tag=tag, auc=auc, acc=acc, rec=rec, cm=cm,
                prob=prob, true=y)

# 单模型训练集表现（可理解为 upper‑bound）
xgb_single = XGBClassifier(**best_xgb).fit(X_tr_s, y_tr)
prob_tr    = xgb_single.predict_proba(X_tr_s)[:,1]

train_res = evaluate(y_tr, prob_tr, best_thr, "Train")
valid_res = evaluate(y_val, prob_val_cal, best_thr, "Valid")
test_res  = evaluate(y_te,  prob_te_cal,  best_thr, "Test")

# ---------------- 10. Confusion & ROC ------
for res in [train_res, valid_res, test_res]:
    ConfusionMatrixDisplay(res["cm"],
        display_labels=["0 (neg)","1 (pos)"]).plot(cmap="Blues")
    plt.title(f'{res["tag"]} Confusion Matrix'); plt.tight_layout(); plt.show()

plt.figure(figsize=(6,5))
for res, sty in zip([train_res, valid_res, test_res], ["-","--","-."]):
    fpr, tpr, _ = roc_curve(res["true"], res["prob"])
    plt.plot(fpr, tpr, sty, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
plt.plot([0,1],[0,1],"k:",lw=1)
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC – XGBoost bagging (early‑stop in CV)")
plt.grid(); plt.legend(); plt.tight_layout(); plt.show()

# ---------------- 11. summary -------------
summary = (pd.DataFrame(
    [{"Dataset":r["tag"],"AUC":r["auc"],
      "Accuracy":r["acc"],"Recall₁":r["rec"]}
     for r in [train_res, valid_res, test_res]])
    .set_index("Dataset").round(3))
print("\n=== Performance Summary ===")
print(summary)
```

```python
##############################################################
#  FULL PIPELINE — HalvingRandomSearchCV + early‑stop + Platt
##############################################################

# ---------------- 0. Imports ----------------
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.metrics import (roc_auc_score, accuracy_score, roc_curve,
                             confusion_matrix, ConfusionMatrixDisplay,
                             balanced_accuracy_score, f1_score)
from sklearn.experimental import enable_halving_search_cv  # noqa
from sklearn.model_selection import HalvingRandomSearchCV
import joblib

import xgboost as xgb

# ---------------- 1. Split 7:2:1 ------------
X = df.drop(columns=['final'])
y = (df['final'] == 'A').astype(int)
X_tr, X_tmp, y_tr, y_tmp = train_test_split(X, y, test_size=0.3,
                                            stratify=y, random_state=42)
X_val, X_te, y_val, y_te = train_test_split(X_tmp, y_tmp, test_size=1/3,
                                            stratify=y_tmp, random_state=42)

# ---------------- 2. logistic top‑24 --------
scaler_lr = StandardScaler().fit(X_tr)
logreg    = LogisticRegression(max_iter=2000, solver="liblinear")
logreg.fit(scaler_lr.transform(X_tr), y_tr)
top_idx   = np.argsort(np.abs(logreg.coef_[0]))[-24:]
TOP24     = X_tr.columns[top_idx]

# -------- 3. feature weighter ---------------
class FeatureWeighter(BaseEstimator, TransformerMixin):
    def __init__(self, cols, alpha=2.0):
        self.cols, self.alpha = cols, alpha
    def fit(self, X, y=None): return self
    def transform(self, X):
        Xw = X.copy()
        Xw[self.cols] *= self.alpha
        return Xw

# -------- 4. Pipeline -----------------------
xgb_base = XGBClassifier(
    objective="binary:logistic",
    eval_metric="auc",
    random_state=42,
    n_jobs=1,                 # 交给外层并行
    tree_method="hist"        # CPU; 改 'hist', device='cuda' 可用 GPU
)

pipe = Pipeline([
    ("weight", FeatureWeighter(cols=TOP24)),
    ("scale",  StandardScaler()),
    ("xgb",    xgb_base)
])

# -------- 5. param_grid ---------------------
pos_ratio = (y_tr == 1).mean()
spw0      = (1-pos_ratio)/pos_ratio
param_grid = {
    "weight__alpha":         [1.5, 2.0],
    "xgb__n_estimators":     [400, 600],       # ←★ 适当缩小
    "xgb__learning_rate":    [0.001],
    "xgb__max_depth":        [2, 3],
    "xgb__min_child_weight": [5],
    "xgb__gamma":            [0, 1],
    "xgb__subsample":        [0.8],
    "xgb__colsample_bytree": [0.8],
    "xgb__reg_alpha":        [0.2, 0.5, 0.8,1, 1.5],
    "xgb__reg_lambda":       [10, 50],
    "xgb__scale_pos_weight": [spw0 ,spw0*1.5, spw0 * 2.0 ]
}

search = HalvingRandomSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    n_candidates=60, factor=3,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring="roc_auc",
    resource="n_samples",
    max_resources="auto",
    verbose=3, n_jobs=-1, random_state=42
)

# ★★★  这里不再传早停参数，只保留静默输出  ★★★
fit_params = {
    "xgb__verbose": False                # 如要看每棵树的 eval，可删掉这行
}

with joblib.parallel_backend("loky", inner_max_num_threads=1):
    search.fit(X_tr, y_tr, **fit_params)

best_alpha = search.best_params_["weight__alpha"]
best_xgb   = {k.split("__")[1]: v for k, v in search.best_params_.items()
              if k.startswith("xgb__")}

# ---------------- 6. Pre‑processing ---------
def weight(df, alpha):
    dfc = df.copy()
    dfc[TOP24] *= alpha
    return dfc

X_tv   = pd.concat([X_tr, X_val]); y_tv = pd.concat([y_tr, y_val])
scaler = StandardScaler().fit(weight(X_tv, best_alpha))
X_tv_s = scaler.transform(weight(X_tv, best_alpha))
X_val_s= scaler.transform(weight(X_val, best_alpha))
X_te_s = scaler.transform(weight(X_te, best_alpha))
X_tr_s = scaler.transform(weight(X_tr, best_alpha))

# ---------------- 7. bagging (5×) ----------
prob_val = np.zeros(len(X_val))
prob_te  = np.zeros(len(X_te))
for seed in range(5):
    boot   = np.random.choice(len(X_tv_s), len(X_tv_s), replace=True)
    model  = XGBClassifier(**best_xgb, random_state=seed,
                           objective="binary:logistic",
                           eval_metric="auc",
                           n_jobs=1, tree_method="hist")
    model.fit(X_tv_s[boot], y_tv.values[boot])
    prob_val += model.predict_proba(X_val_s)[:,1]
    prob_te  += model.predict_proba(X_te_s)[:,1]
prob_val /= 5; prob_te /= 5

# ---------------- 8. Platt scaling ----------
platt = LogisticRegression(solver="lbfgs").fit(
    prob_val.reshape(-1,1), y_val
)
prob_val_cal = platt.predict_proba(prob_val.reshape(-1,1))[:,1]
prob_te_cal  = platt.predict_proba(prob_te.reshape(-1,1))[:,1]

# choose threshold that maximises balanced accuracy on valid
ths = np.linspace(0.1, 0.9, 81)

costs = []
for t in ths:
    pred = (prob_val_cal >= t).astype(int)
    cm = confusion_matrix(y_val, pred)
    tn, fp, fn, tp = cm.ravel()
    cost = 1 * fp + 1.8 * fn  #
    costs.append(cost)

best_thr = ths[np.argmin(costs)]
print(f"Cost-sensitive best threshold = {best_thr:.2f}")


# ---------------- 9. evaluation ------------
def evaluate(y, prob, thr, tag):
    pred = (prob >= thr).astype(int)
    auc  = roc_auc_score(y, prob)
    acc  = accuracy_score(y, pred)
    cm   = confusion_matrix(y, pred)
    rec  = cm[1,1] / (cm[1,1]+cm[1,0])
    print(f"{tag:<6}| AUC={auc:.3f} | ACC={acc:.3f} | Recall₁={rec:.3f}")
    return dict(tag=tag, auc=auc, acc=acc, rec=rec, cm=cm,
                prob=prob, true=y)

# 单模型训练集表现（可理解为 upper‑bound）
# 加入列名后转成 DataFrame
X_tr_named = pd.DataFrame(X_tr_s, columns=weight(X_tr, best_alpha).columns)
xgb_single = XGBClassifier(**best_xgb).fit(X_tr_named, y_tr)
prob_tr_raw = xgb_single.predict_proba(X_tr_s)[:, 1]

# 💡 用验证集训练好的 platt 做校准（不能 refit！）
prob_tr_cal = platt.predict_proba(prob_tr_raw.reshape(-1, 1))[:, 1]

# 统一使用校准后的概率 + 同一个 best_thr
train_res = evaluate(y_tr, prob_tr_cal, best_thr, "Train")
valid_res = evaluate(y_val, prob_val_cal, best_thr, "Valid")
test_res  = evaluate(y_te,  prob_te_cal,  best_thr, "Test")

# ---------------- 10. Confusion & ROC ------
for res in [train_res, valid_res, test_res]:
    ConfusionMatrixDisplay(res["cm"],
        display_labels=["0 (neg)","1 (pos)"]).plot(cmap="Blues")
    plt.title(f'{res["tag"]} Confusion Matrix'); plt.tight_layout(); plt.show()

plt.figure(figsize=(6,5))
for res, sty in zip([train_res, valid_res, test_res], ["-","--","-."]):
    fpr, tpr, _ = roc_curve(res["true"], res["prob"])
    plt.plot(fpr, tpr, sty, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
plt.plot([0,1],[0,1],"k:",lw=1)
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC – XGBoost bagging (early‑stop in CV)")
plt.grid(); plt.legend(); plt.tight_layout(); plt.show()

# ---------------- 11. summary -------------
summary = (pd.DataFrame(
    [{"Dataset":r["tag"],"AUC":r["auc"],
      "Accuracy":r["acc"],"Recall₁":r["rec"]}
     for r in [train_res, valid_res, test_res]])
    .set_index("Dataset").round(3))
print("\n=== Performance Summary ===")
print(summary)
```

```python
# ✅ Full pipeline: Stacking + Balanced Bagging + Focal Loss for tabular binary classification

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (roc_auc_score, accuracy_score, roc_curve,
                             confusion_matrix, ConfusionMatrixDisplay,
                             balanced_accuracy_score, f1_score)
from sklearn.ensemble import StackingClassifier
from sklearn.utils import shuffle
from imblearn.ensemble import BalancedBaggingClassifier
from imblearn.over_sampling import SMOTE
import xgboost as xgb

# ---------------- 1. Split Data -------------
X = df.drop(columns=['final'])
y = (df['final'] == 'A').astype(int)
X_tr, X_tmp, y_tr, y_tmp = train_test_split(X, y, test_size=0.3,
                                            stratify=y, random_state=42)
X_val, X_te, y_val, y_te = train_test_split(X_tmp, y_tmp, test_size=1/3,
                                            stratify=y_tmp, random_state=42)

# ---------------- 2. Combine Train+Val ------
X_meta = pd.concat([X_tr, X_val])
y_meta = pd.concat([y_tr, y_val])

# ---------------- 3. Standardize -------------
scaler = StandardScaler().fit(X_meta)
X_tr_s = scaler.transform(X_tr)
X_val_s = scaler.transform(X_val)
X_te_s  = scaler.transform(X_te)
X_meta_s = scaler.transform(X_meta)

# ---------------- 4. Focal Loss for XGBoost ----
def focal_loss(preds, dtrain):
    gamma = 2.0
    alpha = 0.25
    labels = dtrain.get_label()
    preds = 1.0 / (1.0 + np.exp(-preds))
    grad = alpha * (labels - preds) * ((1 - preds) ** gamma)
    hess = alpha * ((1 - preds) ** (gamma - 1)) * preds * (1 - preds)
    return grad, hess

def focal_loss_metric(preds, dtrain):
    labels = dtrain.get_label()
    preds = 1.0 / (1.0 + np.exp(-preds))
    auc = roc_auc_score(labels, preds)
    return 'focal_auc', auc

# ---------------- 5. Base Learners ----------------
xgb_focal = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=3,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    tree_method='hist',
    random_state=42
)

# Wrap with BalancedBagging (auto-SMOTE)
bagged_xgb = BalancedBaggingClassifier(
    estimator=xgb_focal,
    sampling_strategy='auto',
    replacement=False,
    random_state=42,
    n_estimators=5,
    n_jobs=-1
)

# Logistic regression as a second base learner
base_lr = LogisticRegression(max_iter=2000, solver='liblinear')

# ---------------- 6. Meta Learner ----------------
meta_learner = LogisticRegression(max_iter=1000)

# ---------------- 7. Build Stacking Model --------
stack_model = StackingClassifier(
    estimators=[
        ('xgb_bag', bagged_xgb),
        ('lr', base_lr)
    ],
    final_estimator=meta_learner,
    passthrough=True,
    cv=5,
    n_jobs=-1
)

# ---------------- 8. Fit Model ----------------
from sklearn.calibration import CalibratedClassifierCV
stack_model_calibrated = CalibratedClassifierCV(stack_model, method='sigmoid', cv=3)
stack_model_calibrated.fit(X_meta_s, y_meta)


# ---------------- 9. Predict Probabilities -----


prob_tr = stack_model_calibrated.predict_proba(X_tr_s)[:, 1]
prob_val = stack_model_calibrated.predict_proba(X_val_s)[:, 1]
prob_te = stack_model_calibrated.predict_proba(X_te_s)[:, 1]


# ---------------- 10. Threshold Scanning (recall-aware) -------
# choose threshold that maximises balanced accuracy on valid
ths = np.linspace(0.1, 0.9, 81)
costs = []
for t in ths:
    pred = (prob_val >= t).astype(int)  # 🟢 现在用的是校准后的概率
    cm = confusion_matrix(y_val, pred)
    tn, fp, fn, tp = cm.ravel()
    cost = 1 * fp + 5 * fn
    costs.append(cost)
best_thr = ths[np.argmin(costs)]


# ---------------- 11. Evaluation ---------------
def evaluate(y, prob, thr, tag):
    pred = (prob >= thr).astype(int)
    auc = roc_auc_score(y, prob)
    acc = accuracy_score(y, pred)
    cm = confusion_matrix(y, pred)
    rec = cm[1,1] / (cm[1,1] + cm[1,0]) if (cm[1,1]+cm[1,0]) else 0.0
    f1  = f1_score(y, pred)
    return dict(tag=tag, auc=auc, acc=acc, rec=rec, f1=f1, cm=cm, prob=prob, true=y)

res_tr = evaluate(y_tr, prob_tr, best_thr, "Train")
res_val = evaluate(y_val, prob_val, best_thr, "Valid")
res_te = evaluate(y_te, prob_te, best_thr, "Test")

# ---------------- 12. Confusion Matrices -------
for res in [res_tr, res_val, res_te]:
    ConfusionMatrixDisplay(res["cm"], display_labels=["0", "1"]).plot(cmap="Blues")
    plt.title(f'{res["tag"]} Confusion Matrix')
    plt.tight_layout()
    plt.show()

# ---------------- 13. ROC Curves ---------------
plt.figure(figsize=(6,5))
for res, style in zip([res_tr, res_val, res_te], ["-", "--", "-."]):
    fpr, tpr, _ = roc_curve(res["true"], res["prob"])
    plt.plot(fpr, tpr, style, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
plt.plot([0,1],[0,1],'k:', lw=1)
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC – Stacking + Balanced Bagging + Focal Loss")
plt.legend(loc="lower right", fontsize=10, frameon=True)
plt.grid(); plt.legend(); plt.tight_layout(); plt.show()


# ---------------- 14. Summary ------------------
summary = pd.DataFrame([
    {"Dataset": r["tag"], "AUC": r["auc"], "Accuracy": r["acc"], "Recall₁": r["rec"], "F1": r["f1"]}
    for r in [res_tr, res_val, res_te]
]).set_index("Dataset").round(3)

print("\n=== Final Model Performance Summary ===")
print(summary)
```

```python
##############################################################
#  retrain()  —  快速重训 + 评估 + 可视化                    #
##############################################################
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.metrics import (roc_auc_score, accuracy_score, roc_curve,
                             confusion_matrix, ConfusionMatrixDisplay,
                             balanced_accuracy_score)
from sklearn.experimental import enable_halving_search_cv  # noqa
from sklearn.model_selection import HalvingRandomSearchCV
import joblib, xgboost
from xgboost import XGBClassifier

def retrain(df, drop_cols=None, keep_cols=None, custom_grid=None, seed=42):
    # ---------- A. 预处理列 ----------
    X = df.drop(columns=['final'])
    if keep_cols is not None:
        X = X[keep_cols]
    if drop_cols is not None:
        X = X.drop(columns=drop_cols)
    y = (df['final'] == 'T').astype(int)

    X_tr, X_tmp, y_tr, y_tmp = train_test_split(
        X, y, test_size=0.3, stratify=y, random_state=seed)
    X_val, X_te, y_val, y_te = train_test_split(
        X_tmp, y_tmp, test_size=1/3, stratify=y_tmp, random_state=seed)

    # ---------- B. Logistic Top‑24 ----------
    scaler_lr = StandardScaler().fit(X_tr)
    lr = LogisticRegression(max_iter=2000, solver="liblinear").fit(
        scaler_lr.transform(X_tr), y_tr)
    top_idx = np.argsort(np.abs(lr.coef_[0]))[-24:]
    TOP24 = X_tr.columns[top_idx].tolist()

    # ---------- C. Pipeline ----------
    class FeatureWeighter(BaseEstimator, TransformerMixin):
        def __init__(self, cols, alpha=2.0):
            self.cols, self.alpha = cols, alpha
        def fit(self, X, y=None): return self
        def transform(self, X):
            if isinstance(X, pd.DataFrame):
                Xw = X.copy(); Xw[self.cols] *= self.alpha; return Xw
            # 若传入 ndarray，直接返回，避免 IndexError
            return X

    xgb_base = XGBClassifier(
        objective="binary:logistic", eval_metric="auc",
        random_state=seed, n_jobs=1, tree_method="hist")

    pipe = Pipeline([
        ("weight", FeatureWeighter(cols=TOP24)),
        ("scale",  StandardScaler()),
        ("xgb",    xgb_base)
    ])

    # ---------- D. 超参网格 ----------
    pos_ratio = (y_tr == 1).mean()
    spw0 = (1 - pos_ratio) / pos_ratio
    param_grid = {
        "weight__alpha":         [1.5, 2.0],
        "xgb__n_estimators":     [400, 600],
        "xgb__learning_rate":    [0.01],
        "xgb__max_depth":        [2, 3],
        "xgb__min_child_weight": [5, 10],
        "xgb__gamma":            [0, 1],
        "xgb__subsample":        [0.8],
        "xgb__colsample_bytree": [0.8],
        "xgb__reg_alpha":        [1, 5],
        "xgb__reg_lambda":       [10, 50],
        "xgb__scale_pos_weight": [spw0*1.5, spw0*2, spw0*3]
    }
    if custom_grid:
        param_grid.update(custom_grid)

    search = HalvingRandomSearchCV(
        pipe, param_grid, n_candidates=60, factor=3,
        cv=StratifiedKFold(5, shuffle=True, random_state=seed),
        scoring="roc_auc", resource="n_samples", max_resources="auto",
        verbose=3, n_jobs=-1, random_state=seed)

    with joblib.parallel_backend("loky", inner_max_num_threads=1):
        search.fit(X_tr, y_tr, **{"xgb__verbose": False})

    best_alpha = search.best_params_["weight__alpha"]
    best_xgb = {k.split("__")[1]: v for k, v in search.best_params_.items()
                if k.startswith("xgb__")}

    # ---------- E. bagging + Platt ----------
    def w(df): d = df.copy(); d[TOP24] *= best_alpha; return d
    scaler = StandardScaler().fit(w(pd.concat([X_tr, X_val])))
    tr_s, val_s, te_s = [scaler.transform(w(d)) for d in (X_tr, X_val, X_te)]
    tv_s = scaler.transform(w(pd.concat([X_tr, X_val]))); y_tv = pd.concat([y_tr, y_val])

    prob_val = np.zeros(len(X_val)); prob_te = np.zeros(len(X_te))
    for s in range(5):
        boot = np.random.choice(len(tv_s), len(tv_s), True)
        mdl = XGBClassifier(**best_xgb, random_state=s,
                            objective="binary:logistic",
                            eval_metric="auc", n_jobs=1,
                            tree_method="hist").fit(tv_s[boot], y_tv.values[boot])
        prob_val += mdl.predict_proba(val_s)[:, 1]
        prob_te  += mdl.predict_proba(te_s)[:, 1]
    prob_val /= 5; prob_te /= 5

    platt = LogisticRegression(solver="lbfgs").fit(prob_val.reshape(-1, 1), y_val)
    prob_val_cal = platt.predict_proba(prob_val.reshape(-1, 1))[:, 1]
    prob_te_cal  = platt.predict_proba(prob_te.reshape(-1, 1))[:, 1]

    ths = np.linspace(0.1, 0.9, 81)
    thr = ths[int(np.argmax([balanced_accuracy_score(y_val, (prob_val_cal >= t))
                             for t in ths]))]
    print(f"\n◎ Balanced‑ACC threshold = {thr:.2f}")

    def eva(y, p, tag):
        pred = (p >= thr).astype(int)
        auc = roc_auc_score(y, p); acc = accuracy_score(y, pred)
        cm = confusion_matrix(y, pred)
        rec = cm[1, 1] / (cm[1, 1] + cm[1, 0])
        print(f"{tag:<6}| AUC={auc:.3f}|ACC={acc:.3f}|Recall₁={rec:.3f}")
        return dict(tag=tag, auc=auc, acc=acc, rec=rec, cm=cm, prob=p, true=y)

    # ★ 传入原始 DataFrame，避免 IndexError
    tr_res = eva(y_tr, search.best_estimator_.predict_proba(X_tr)[:, 1], "Train")
    va_res = eva(y_val, prob_val_cal, "Valid")
    te_res = eva(y_te,  prob_te_cal,  "Test")

    # ---------- F. 可视化 ----------
    for r in (tr_res, va_res, te_res):
        ConfusionMatrixDisplay(r["cm"], display_labels=["0", "1"]).plot(cmap="Blues")
        plt.title(f'{r["tag"]} Confusion Matrix'); plt.tight_layout(); plt.show()

    plt.figure(figsize=(5, 4))
    for r, sty in zip((tr_res, va_res, te_res), ("-", "--", "-.")):
        fpr, tpr, _ = roc_curve(r["true"], r["prob"])
        plt.plot(fpr, tpr, sty, label=f'{r["tag"]} AUC={r["auc"]:.2f}')
    plt.plot([0, 1], [0, 1], "k:", lw=1)
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.grid(); plt.legend(); plt.tight_layout(); plt.show()

    return (pd.DataFrame([{"Dataset": r["tag"], "AUC": r["auc"],
                           "ACC": r["acc"], "Recall": r["rec"]}
                          for r in (tr_res, va_res, te_res)])
            .set_index("Dataset").round(3))
```

```python
import numpy as np
import pandas as pd
from sklearn.mixture import BayesianGaussianMixture
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score
import itertools

# 假设你已经做过 7:2:1 划分，得到了：
# X_train, y_train;  X_valid, y_valid

# 要搜索的增强超参数空间
param_grid = {
    'n_components': [4, 6, 8],
    'covariance_type': ['diag', 'tied'],
    'factor': [1, 2]
}

results = []

for n_comp, cov_type, factor in itertools.product(
        param_grid['n_components'],
        param_grid['covariance_type'],
        param_grid['factor']
    ):
    # —— 1. 拆分正负类
    X0 = X_train[y_train == 0].values
    X1 = X_train[y_train == 1].values

    # —— 2. 拟合 BGMM
    bgmm0 = BayesianGaussianMixture(
        n_components=n_comp,
        covariance_type=cov_type,
        weight_concentration_prior_type='dirichlet_process',
        random_state=42
    ).fit(X0)
    bgmm1 = BayesianGaussianMixture(
        n_components=n_comp,
        covariance_type=cov_type,
        weight_concentration_prior_type='dirichlet_process',
        random_state=42
    ).fit(X1)

    # —— 3. 采样增强
    X0_new, _ = bgmm0.sample(len(X0) * factor)
    X1_new, _ = bgmm1.sample(len(X1) * factor)

    X_aug = np.vstack([X_train.values, X0_new, X1_new])
    y_aug = np.concatenate([
        y_train.values,
        np.zeros(len(X0_new), dtype=int),
        np.ones(len(X1_new), dtype=int)
    ])

    # —— 4. 训练同一 MLP 配置
    mlp = MLPClassifier(
        hidden_layer_sizes=(64, 32),
        activation='relu',
        solver='adam',
        alpha=0.001,
        early_stopping=True,
        validation_fraction=0.1,
        n_iter_no_change=10,
        max_iter=300,
        random_state=42,
        verbose=False
    )
    mlp.fit(X_aug, y_aug)

    # —— 5. 在 Validation 上评估 AUC
    y_val_prob = mlp.predict_proba(X_valid)[:, 1]
    val_auc = roc_auc_score(y_valid, y_val_prob)

    # 记录结果
    results.append({
        'n_components': n_comp,
        'covariance_type': cov_type,
        'factor': factor,
        'val_auc': val_auc
    })
    print(f"n_comp={n_comp}, cov_type={cov_type}, factor={factor} → Val AUC = {val_auc:.4f}")

```

```python
# 第4块：特征工程类定义 - 与ATN处理方式一致

class FeatureWeighter(BaseEstimator, TransformerMixin):
    """与您之前ATN代码一致的特征加权器"""
    def __init__(self, cols, alpha=2.0):
        self.cols = cols
        self.alpha = alpha

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        Xw = X.copy()
        if isinstance(Xw, pd.DataFrame):
            # 对指定列进行加权
            Xw[self.cols] *= self.alpha
        return Xw

def select_top_features(X_train, y_train, top_k=24):
    """使用逻辑回归选择top-k重要特征，与ATN代码一致"""

    # 标准化
    scaler_lr = StandardScaler().fit(X_train)
    X_train_scaled = scaler_lr.transform(X_train)

    # 逻辑回归
    logreg = LogisticRegression(max_iter=2000, solver="liblinear", random_state=42)
    logreg.fit(X_train_scaled, y_train)

    # 获取top-k特征（按系数绝对值）
    coef_abs = np.abs(logreg.coef_[0])
    top_idx = np.argsort(coef_abs)[-top_k:]
    TOP_FEATURES = X_train.columns[top_idx]

    print(f"逻辑回归选出的前{top_k}重要特征:")
    feature_importance = list(zip(TOP_FEATURES, coef_abs[top_idx]))
    feature_importance.sort(key=lambda x: x[1], reverse=True)

    for i, (feature, importance) in enumerate(feature_importance):
        print(f"{i+1:2d}. {feature}: {importance:.4f}")

    return TOP_FEATURES.tolist()
```

```python
# 第6块：建模管道 - 与ATN处理方式一致

def create_modeling_pipeline(top_features, alpha=2.0):
    """创建与ATN一致的建模管道"""

    # 计算不平衡权重，与您ATN代码一致
    pos_ratio = y_train.mean()
    spw0 = (1 - pos_ratio) / pos_ratio

    # XGBoost基础模型
    xgb_base = XGBClassifier(
        objective="binary:logistic",
        eval_metric="auc",
        use_label_encoder=False,
        random_state=42,
        n_jobs=1,  # 与您ATN代码一致
        tree_method="hist"
    )

    # 创建管道：特征加权 → 标准化 → XGBoost （与ATN一致）
    pipeline = Pipeline([
        ("weight", FeatureWeighter(cols=top_features, alpha=alpha)),
        ("scale", StandardScaler()),
        ("xgb", xgb_base)
    ])

    return pipeline, spw0

def get_param_grid(spw0):
    """获取参数搜索空间，与ATN代码参数范围一致"""

    param_grid = {
        # 特征加权参数（与您ATN代码一致）
        "weight__alpha": [1.5, 2.0, 2.5],

        # XGBoost参数（参考您ATN代码的范围）
        "xgb__n_estimators": [400, 600, 800],
        "xgb__learning_rate": [0.001, 0.01, 0.05],
        "xgb__max_depth": [2, 3, 4],
        "xgb__min_child_weight": [5, 10],
        "xgb__gamma": [0, 1],
        "xgb__subsample": [0.8],
        "xgb__colsample_bytree": [0.8],
        "xgb__reg_alpha": [0.2, 0.5, 1, 1.5],
        "xgb__reg_lambda": [10, 50],
        "xgb__scale_pos_weight": [spw0, spw0*1.5, spw0*2.0]
    }

    return param_grid

# 创建管道（使用已选择的TOP24特征）
print("创建建模管道...")
pipeline, spw0 = create_modeling_pipeline(TOP24_FEATURES, alpha=2.0)

print("建模管道创建成功！")
print(f"\n管道步骤:")
for i, (name, step) in enumerate(pipeline.named_steps.items()):
    print(f"{i+1}. {name}: {type(step).__name__}")

print(f"\n将对以下{len(TOP24_FEATURES)}个特征进行加权:")
print(TOP24_FEATURES)

# 获取参数网格
param_grid = get_param_grid(spw0)
print(f"\n参数搜索空间:")
total_combinations = np.prod([len(v) for v in param_grid.values()])
print(f"总参数组合数: {total_combinations:,}")

# 显示关键参数设置
print(f"\n关键参数设置:")
print(f"  正类比例: {y_train.mean():.3f}")
print(f"  scale_pos_weight基准值: {spw0:.3f}")
print(f"  alpha权重范围: {param_grid['weight__alpha']}")
print(f"  n_estimators范围: {param_grid['xgb__n_estimators']}")
print(f"  学习率范围: {param_grid['xgb__learning_rate']}")

print("\n建模管道准备完成！")
```

```python
# 第7块：模型训练和超参数搜索

def train_model_with_search(X_train, X_val, y_train, y_val, pipeline, param_grid, verbose=True):
    """使用HalvingRandomSearchCV进行模型训练和超参数搜索"""

    print("开始超参数搜索...")
    print(f"参数空间大小: {np.prod([len(v) for v in param_grid.values()]):,}")

    # 设置搜索器
    search = HalvingRandomSearchCV(
        estimator=pipeline,
        param_distributions=param_grid,
        n_candidates=60,  # 与您ATN代码一致
        factor=3,         # 与您ATN代码一致
        cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
        scoring="roc_auc",
        resource="n_samples",
        max_resources="auto",
        verbose=2 if verbose else 0,
        n_jobs=-1,
        random_state=42
    )

    # 训练参数（静默XGBoost输出）
    fit_params = {
        "xgb__verbose": False
    }

    # 执行搜索
    print("正在进行参数搜索，这可能需要几分钟...")
    with joblib.parallel_backend("loky", inner_max_num_threads=1):
        search.fit(X_train, y_train, **fit_params)

    print(f"\n搜索完成！")
    print(f"最佳CV AUC: {search.best_score_:.4f}")
    print(f"最佳参数:")
    for param, value in search.best_params_.items():
        print(f"  {param}: {value}")

    return search

# 开始训练
print("=" * 50)
print("开始Normal vs Disease模型训练")
print("=" * 50)

# 检查必要变量是否存在
required_vars = ['X_train', 'y_train', 'X_val', 'y_val', 'pipeline', 'param_grid']
missing_vars = [var for var in required_vars if var not in globals()]

if missing_vars:
    print(f"错误：缺少必要变量: {missing_vars}")
    print("请确保已运行前面的所有代码块")
else:
    print(f"训练集: {X_train.shape[0]} 样本")
    print(f"验证集: {X_val.shape[0]} 样本")
    print(f"特征数: {X_train.shape[1]}")
    print(f"重点特征数: {len(TOP24_FEATURES)}")

    # 执行训练
    search_result = train_model_with_search(
        X_train, X_val, y_train, y_val,
        pipeline, param_grid, verbose=True
    )

    print(f"\n模型训练完成！")

    # 保存最佳模型
    best_model = search_result.best_estimator_
    print("最佳模型已保存到变量 'best_model'")
```

```python
# 第一步：数据质量检查和调试
def diagnose_data_quality(X_train, y_train, X_val, y_val):
    """全面诊断数据质量问题"""
    print("="*50)
    print("数据质量诊断")
    print("="*50)

    # 1. 基本信息
    print(f"训练集形状: {X_train.shape}")
    print(f"验证集形状: {X_val.shape}")
    print(f"标签分布:")
    print(f"  训练集: {dict(y_train.value_counts())}")
    print(f"  验证集: {dict(y_val.value_counts())}")

    # 2. 检查缺失值
    train_missing = X_train.isnull().sum().sum()
    val_missing = X_val.isnull().sum().sum()
    print(f"\n缺失值检查:")
    print(f"  训练集缺失值总数: {train_missing}")
    print(f"  验证集缺失值总数: {val_missing}")

    if train_missing > 0:
        print(f"  训练集缺失值分布:")
        missing_cols = X_train.isnull().sum()
        for col in missing_cols[missing_cols > 0].index:
            print(f"    {col}: {missing_cols[col]} ({missing_cols[col]/len(X_train)*100:.1f}%)")

    # 3. 检查无限值
    train_inf = np.isinf(X_train.select_dtypes(include=[np.number])).sum().sum()
    val_inf = np.isinf(X_val.select_dtypes(include=[np.number])).sum().sum()
    print(f"\n无限值检查:")
    print(f"  训练集无限值总数: {train_inf}")
    print(f"  验证集无限值总数: {val_inf}")

    # 4. 检查特征方差
    feature_vars = X_train.var()
    zero_var_features = feature_vars[feature_vars == 0].index.tolist()
    low_var_features = feature_vars[(feature_vars > 0) & (feature_vars < 1e-10)].index.tolist()

    print(f"\n特征方差检查:")
    print(f"  零方差特征数: {len(zero_var_features)}")
    print(f"  低方差特征数: {len(low_var_features)}")

    if zero_var_features:
        print(f"  零方差特征: {zero_var_features[:10]}...")  # 只显示前10个

    # 5. 检查特征分布
    print(f"\n特征统计:")
    print(f"  数值特征数: {X_train.select_dtypes(include=[np.number]).shape[1]}")
    print(f"  特征均值范围: [{X_train.mean().min():.3f}, {X_train.mean().max():.3f}]")
    print(f"  特征标准差范围: [{X_train.std().min():.3f}, {X_train.std().max():.3f}]")

    # 6. 类别平衡分析
    pos_ratio = y_train.mean()
    print(f"\n类别平衡分析:")
    print(f"  正类比例: {pos_ratio:.3f}")
    print(f"  负类比例: {1-pos_ratio:.3f}")
    print(f"  不平衡比率: {(1-pos_ratio)/pos_ratio:.2f}:1")

    return {
        'missing_values': train_missing > 0,
        'infinite_values': train_inf > 0,
        'zero_variance': len(zero_var_features) > 0,
        'severe_imbalance': pos_ratio > 0.9 or pos_ratio < 0.1,
        'zero_var_features': zero_var_features,
        'low_var_features': low_var_features
    }

# 第二步：数据预处理改进
def improved_data_preprocessing(X_train, X_val, X_test, y_train):
    """改进的数据预处理"""
    print("\n" + "="*50)
    print("改进数据预处理")
    print("="*50)

    # 创建副本
    X_train_clean = X_train.copy()
    X_val_clean = X_val.copy()
    X_test_clean = X_test.copy()

    # 1. 处理缺失值
    if X_train_clean.isnull().sum().sum() > 0:
        print("处理缺失值...")
        # 数值型特征用中位数填充
        numeric_cols = X_train_clean.select_dtypes(include=[np.number]).columns
        for col in numeric_cols:
            if X_train_clean[col].isnull().sum() > 0:
                median_val = X_train_clean[col].median()
                X_train_clean[col].fillna(median_val, inplace=True)
                X_val_clean[col].fillna(median_val, inplace=True)
                X_test_clean[col].fillna(median_val, inplace=True)
                print(f"  {col}: 用{median_val}填充")

    # 2. 处理无限值
    for df in [X_train_clean, X_val_clean, X_test_clean]:
        df.replace([np.inf, -np.inf], np.nan, inplace=True)
        if df.isnull().sum().sum() > 0:
            df.fillna(df.median(), inplace=True)

    # 3. 移除零方差特征
    feature_vars = X_train_clean.var()
    zero_var_features = feature_vars[feature_vars == 0].index.tolist()

    if zero_var_features:
        print(f"移除{len(zero_var_features)}个零方差特征")
        X_train_clean = X_train_clean.drop(columns=zero_var_features)
        X_val_clean = X_val_clean.drop(columns=zero_var_features)
        X_test_clean = X_test_clean.drop(columns=zero_var_features)

    # 4. 特征缩放（强制使用RobustScaler，对异常值更鲁棒）
    print("应用RobustScaler...")
    scaler = RobustScaler()
    X_train_scaled = pd.DataFrame(
        scaler.fit_transform(X_train_clean),
        columns=X_train_clean.columns,
        index=X_train_clean.index
    )
    X_val_scaled = pd.DataFrame(
        scaler.transform(X_val_clean),
        columns=X_val_clean.columns,
        index=X_val_clean.index
    )
    X_test_scaled = pd.DataFrame(
        scaler.transform(X_test_clean),
        columns=X_test_clean.columns,
        index=X_test_clean.index
    )

    print(f"预处理完成！最终特征数: {X_train_scaled.shape[1]}")

    return X_train_scaled, X_val_scaled, X_test_scaled, scaler

# 第三步：改进的特征选择
def improved_feature_selection(X_train, y_train, top_k=20):
    """改进的特征选择方法"""
    print("\n" + "="*50)
    print("改进特征选择")
    print("="*50)

    feature_scores = {}

    # 方法1: 互信息
    print("1. 计算互信息...")
    mi_scores = mutual_info_classif(X_train, y_train, random_state=42)
    feature_scores['mutual_info'] = dict(zip(X_train.columns, mi_scores))

    # 方法2: 卡方检验（需要非负特征）
    print("2. 计算卡方统计量...")
    # 将特征转换为非负（通过MinMaxScaler）
    min_max_scaler = MinMaxScaler()
    X_train_positive = min_max_scaler.fit_transform(X_train)
    chi2_scores, _ = chi2(X_train_positive, y_train)
    feature_scores['chi2'] = dict(zip(X_train.columns, chi2_scores))

    # 方法3: 随机森林重要性
    print("3. 计算随机森林重要性...")
    rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    feature_scores['rf_importance'] = dict(zip(X_train.columns, rf.feature_importances_))

    # 方法4: L1正则化逻辑回归
    print("4. 计算L1逻辑回归系数...")
    lr_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=0.1, random_state=42)
    lr_l1.fit(X_train, y_train)
    feature_scores['lr_l1'] = dict(zip(X_train.columns, np.abs(lr_l1.coef_[0])))

    # 综合评分（排名平均）
    print("5. 综合特征评分...")
    feature_rankings = {}

    for feature in X_train.columns:
        rankings = []
        for method, scores in feature_scores.items():
            sorted_features = sorted(scores.items(), key=lambda x: x[1], reverse=True)
            rank = next(i for i, (feat, _) in enumerate(sorted_features) if feat == feature)
            rankings.append(rank)
        feature_rankings[feature] = np.mean(rankings)

    # 选择top-k特征
    top_features = sorted(feature_rankings.items(), key=lambda x: x[1])[:top_k]
    top_feature_names = [feat for feat, _ in top_features]

    print(f"\n选出的top-{top_k}特征:")
    for i, (feature, avg_rank) in enumerate(top_features):
        print(f"{i+1:2d}. {feature}: 平均排名 {avg_rank:.1f}")

    return top_feature_names, feature_scores

# 第四步：改进的模型定义
def create_improved_pipeline(selected_features):
    """创建改进的模型管道"""
    print("\n" + "="*50)
    print("创建改进模型管道")
    print("="*50)

    # 定义多个基础模型
    models = {
        'xgb': XGBClassifier(
            objective="binary:logistic",
            eval_metric="auc",
            use_label_encoder=False,
            random_state=42,
            tree_method="hist"
        ),
        'lgb': LGBMClassifier(
            objective='binary',
            metric='auc',
            random_state=42,
            verbose=-1
        ),
        'rf': RandomForestClassifier(
            random_state=42,
            n_jobs=-1
        )
    }

    # 特征选择器
    feature_selector = ColumnTransformer([
        ('selected', 'passthrough', selected_features)
    ], remainder='drop')

    # 为每个模型创建管道
    pipelines = {}
    for name, model in models.items():
        pipeline = Pipeline([
            ('feature_select', feature_selector),
            ('scale', StandardScaler()),
            ('model', model)
        ])
        pipelines[name] = pipeline

    return pipelines

# 第五步：改进的参数搜索
def get_improved_param_grids(y_train):
    """获取改进的参数网格"""

    # 计算类别权重
    pos_ratio = y_train.mean()
    neg_ratio = 1 - pos_ratio
    scale_pos_weight = neg_ratio / pos_ratio

    param_grids = {
        'xgb': {
            'model__n_estimators': [100, 200, 300],
            'model__learning_rate': [0.01, 0.05, 0.1],
            'model__max_depth': [3, 4, 5],
            'model__min_child_weight': [1, 3, 5],
            'model__subsample': [0.8, 0.9, 1.0],
            'model__colsample_bytree': [0.8, 0.9, 1.0],
            'model__scale_pos_weight': [scale_pos_weight, scale_pos_weight*0.5, scale_pos_weight*2.0]
        },
        'lgb': {
            'model__n_estimators': [100, 200, 300],
            'model__learning_rate': [0.01, 0.05, 0.1],
            'model__max_depth': [3, 4, 5],
            'model__min_child_samples': [10, 20, 30],
            'model__subsample': [0.8, 0.9, 1.0],
            'model__colsample_bytree': [0.8, 0.9, 1.0],
            'model__scale_pos_weight': [scale_pos_weight, scale_pos_weight*0.5, scale_pos_weight*2.0]
        },
        'rf': {
            'model__n_estimators': [100, 200, 300],
            'model__max_depth': [5, 10, 15, None],
            'model__min_samples_split': [2, 5, 10],
            'model__min_samples_leaf': [1, 2, 4],
            'model__class_weight': ['balanced', 'balanced_subsample', None]
        }
    }

    return param_grids

# 第六步：简化的模型训练
def train_improved_models(X_train, y_train, pipelines, param_grids):
    """训练改进的模型"""
    print("\n" + "="*50)
    print("训练改进模型")
    print("="*50)

    best_models = {}
    results = {}

    for model_name in pipelines.keys():
        print(f"\n训练 {model_name.upper()} 模型...")

        # 使用较小的搜索空间和更简单的交叉验证
        search = RandomizedSearchCV(
            estimator=pipelines[model_name],
            param_distributions=param_grids[model_name],
            n_iter=20,  # 减少搜索次数
            cv=3,  # 减少交叉验证折数
            scoring='roc_auc',
            random_state=42,
            n_jobs=-1,
            verbose=1
        )

        try:
            search.fit(X_train, y_train)
            best_models[model_name] = search.best_estimator_
            results[model_name] = {
                'best_score': search.best_score_,
                'best_params': search.best_params_
            }
            print(f"{model_name.upper()} 最佳CV AUC: {search.best_score_:.4f}")

        except Exception as e:
            print(f"{model_name.upper()} 训练失败: {str(e)}")
            continue

    return best_models, results

# 执行调试和改进流程
print("开始模型调试和改进流程...")

# 必要的导入
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler
from sklearn.feature_selection import mutual_info_classif, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# 如果有数据，开始诊断
if 'X_train' in globals():
    # 第一步：诊断数据质量
    issues = diagnose_data_quality(X_train, y_train, X_val, y_val)

    # 第二步：数据预处理
    X_train_processed, X_val_processed, X_test_processed, scaler = improved_data_preprocessing(
        X_train, X_val, X_test, y_train
    )

    # 第三步：特征选择
    selected_features, feature_scores = improved_feature_selection(
        X_train_processed, y_train, top_k=20
    )

    # 第四步：创建改进的模型管道
    improved_pipelines = create_improved_pipeline(selected_features)

    # 第五步：获取参数网格
    improved_param_grids = get_improved_param_grids(y_train)

    # 第六步：训练模型
    best_models, training_results = train_improved_models(
        X_train_processed, y_train, improved_pipelines, improved_param_grids
    )

    print("\n" + "="*50)
    print("改进完成！")
    print("="*50)
    print("改进后的变量:")
    print("- X_train_processed, X_val_processed, X_test_processed: 处理后的数据")
    print("- selected_features: 选中的特征列表")
    print("- best_models: 训练好的模型字典")
    print("- training_results: 训练结果")

else:
    print("请先确保前面的数据准备代码已执行!")
```

```python
# 最简单的模型分析解决方案 - 重新训练纯模型
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    precision_recall_curve, roc_curve, auc
)
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

def create_fresh_models():
    """创建全新的简单模型，避开pipeline问题"""

    print("🔄 创建全新的简单模型...")

    # 计算类别权重
    pos_ratio = y_train.mean()
    scale_pos_weight = (1 - pos_ratio) / pos_ratio

    print(f"正类比例: {pos_ratio:.3f}, scale_pos_weight: {scale_pos_weight:.3f}")

    # 创建简单模型（不使用pipeline）
    models = {
        'XGB': XGBClassifier(
            objective='binary:logistic',
            eval_metric='auc',
            n_estimators=200,
            learning_rate=0.05,
            max_depth=4,
            min_child_weight=3,
            subsample=0.8,
            colsample_bytree=0.8,
            scale_pos_weight=scale_pos_weight,
            reg_alpha=1,
            reg_lambda=1,
            random_state=42,
            use_label_encoder=False
        ),

        'LGB': LGBMClassifier(
            objective='binary',
            metric='auc',
            n_estimators=200,
            learning_rate=0.05,
            max_depth=4,
            min_child_samples=10,
            subsample=0.8,
            colsample_bytree=0.8,
            scale_pos_weight=scale_pos_weight,
            reg_alpha=1,
            reg_lambda=1,
            random_state=42,
            verbose=-1
        ),

        'RF': RandomForestClassifier(
            n_estimators=200,
            max_depth=10,
            min_samples_split=5,
            min_samples_leaf=2,
            class_weight='balanced',
            random_state=42,
            n_jobs=-1
        )
    }

    return models

def prepare_data_simple():
    """准备数据（最简单方式）"""

    print("📊 准备数据...")

    # 检查数据类型
    if hasattr(X_train, 'iloc'):
        # DataFrame格式
        print("使用DataFrame格式数据")
        X_tr = X_train.values  # 转换为numpy数组
        X_va = X_val.values
        X_te = X_test.values
        feature_names = X_train.columns.tolist()
    else:
        # 已经是numpy数组
        print("使用numpy数组格式数据")
        X_tr = X_train
        X_va = X_val
        X_te = X_test
        feature_names = [f'feature_{i}' for i in range(X_train.shape[1])]

    print(f"数据形状: Train{X_tr.shape}, Val{X_va.shape}, Test{X_te.shape}")
    print(f"特征数量: {len(feature_names)}")

    # 标准化
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_tr)
    X_val_scaled = scaler.transform(X_va)
    X_test_scaled = scaler.transform(X_te)

    return X_train_scaled, X_val_scaled, X_test_scaled, feature_names

def evaluate_simple_models(models, X_train_scaled, X_val_scaled, X_test_scaled,
                          y_train, y_val, y_test):
    """评估简单模型"""

    print("\n🔍 开始模型评估...")

    results = {}

    for model_name, model in models.items():
        print(f"\n📈 训练和评估 {model_name} 模型...")

        # 训练模型
        model.fit(X_train_scaled, y_train)

        # 在三个数据集上评估
        datasets = {
            'Train': (X_train_scaled, y_train),
            'Validation': (X_val_scaled, y_val),
            'Test': (X_test_scaled, y_test)
        }

        model_results = {}

        for dataset_name, (X, y) in datasets.items():
            # 预测
            y_pred = model.predict(X)
            y_pred_proba = model.predict_proba(X)[:, 1]

            # 计算指标
            auc_roc = roc_auc_score(y, y_pred_proba)
            precision, recall, _ = precision_recall_curve(y, y_pred_proba)
            auc_pr = auc(recall, precision)

            # 混淆矩阵
            cm = confusion_matrix(y, y_pred)
            tn, fp, fn, tp = cm.ravel()

            # 详细指标
            accuracy = (tp + tn) / (tp + tn + fp + fn)
            sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
            precision_score = tp / (tp + fp) if (tp + fp) > 0 else 0
            f1 = 2 * (precision_score * sensitivity) / (precision_score + sensitivity) if (precision_score + sensitivity) > 0 else 0

            model_results[dataset_name] = {
                'auc_roc': auc_roc,
                'auc_pr': auc_pr,
                'accuracy': accuracy,
                'sensitivity': sensitivity,
                'specificity': specificity,
                'precision': precision_score,
                'f1': f1,
                'confusion_matrix': cm
            }

            print(f"  {dataset_name:10}: AUC-ROC={auc_roc:.4f}, AUC-PR={auc_pr:.4f}")

        results[model_name] = model_results

    return results

def print_summary_table(results):
    """打印汇总表"""

    print("\n" + "="*120)
    print("📊 模型性能汇总表")
    print("="*120)

    # 表头
    header = f"{'Model':<8} {'Dataset':<12} {'AUC-ROC':<8} {'AUC-PR':<8} {'Accuracy':<8} {'Sensitivity':<11} {'Specificity':<11} {'Precision':<9} {'F1-Score':<8}"
    print(header)
    print("-" * 120)

    # 数据行
    for model_name, model_results in results.items():
        for dataset_name, metrics in model_results.items():
            row = f"{model_name:<8} {dataset_name:<12} {metrics['auc_roc']:<8.4f} {metrics['auc_pr']:<8.4f} {metrics['accuracy']:<8.4f} {metrics['sensitivity']:<11.4f} {metrics['specificity']:<11.4f} {metrics['precision']:<9.4f} {metrics['f1']:<8.4f}"
            print(row)

def plot_performance_bars(results):
    """绘制性能对比柱状图"""

    models = list(results.keys())
    datasets = ['Train', 'Validation', 'Test']
    metrics = ['auc_roc', 'auc_pr', 'accuracy', 'sensitivity', 'specificity', 'precision']

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.flatten()

    colors = ['skyblue', 'orange', 'lightgreen']

    for i, metric in enumerate(metrics):
        ax = axes[i]

        x = np.arange(len(models))
        width = 0.25

        for j, dataset in enumerate(datasets):
            values = [results[model][dataset][metric] for model in models]
            bars = ax.bar(x + j*width, values, width, label=dataset,
                         color=colors[j], alpha=0.8)

            # 添加数值标签
            for k, bar in enumerate(bars):
                height = bar.get_height()
                ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                       f'{height:.3f}', ha='center', va='bottom', fontsize=9)

        ax.set_xlabel('Models')
        ax.set_ylabel(metric.replace('_', '-').upper())
        ax.set_title(f'{metric.replace("_", "-").upper()} Comparison')
        ax.set_xticks(x + width)
        ax.set_xticklabels(models)
        ax.legend()
        ax.grid(True, alpha=0.3)
        ax.set_ylim(0, 1.1)

    plt.tight_layout()
    plt.show()

def plot_confusion_matrices(results):
    """绘制混淆矩阵"""

    models = list(results.keys())
    datasets = ['Train', 'Validation', 'Test']

    fig, axes = plt.subplots(len(models), len(datasets), figsize=(15, 5*len(models)))

    if len(models) == 1:
        axes = axes.reshape(1, -1)

    for i, model_name in enumerate(models):
        for j, dataset in enumerate(datasets):
            ax = axes[i, j]

            cm = results[model_name][dataset]['confusion_matrix']

            # 绘制热力图
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                       xticklabels=['Normal', 'Disease'],
                       yticklabels=['Normal', 'Disease'],
                       cbar=False)

            ax.set_title(f'{model_name} - {dataset}')
            ax.set_xlabel('Predicted')
            ax.set_ylabel('Actual')

            # 添加百分比
            total = cm.sum()
            for k in range(2):
                for l in range(2):
                    percentage = cm[k, l] / total * 100
                    ax.text(l + 0.5, k + 0.7, f'({percentage:.1f}%)',
                           ha='center', va='center', fontsize=9, color='red')

    plt.tight_layout()
    plt.show()

def analyze_feature_importance(models, feature_names):
    """分析特征重要性"""

    print("\n" + "="*80)
    print("⭐ 特征重要性分析")
    print("="*80)

    for model_name, model in models.items():
        if hasattr(model, 'feature_importances_'):
            importances = model.feature_importances_

            # 创建特征重要性数据
            n_features = min(len(importances), len(feature_names))
            feature_data = list(zip(feature_names[:n_features], importances[:n_features]))
            feature_data.sort(key=lambda x: x[1], reverse=True)

            # 取前10个
            top10 = feature_data[:10]

            print(f"\n{model_name} 模型的Top 10重要特征:")
            print("-" * 60)
            for i, (feature, importance) in enumerate(top10, 1):
                total_gain = importance
                avg_gain = importance / n_features
                print(f"{i:2d}. {feature:30} Total:{total_gain:.4f}  Avg:{avg_gain:.6f}")

            # 绘制特征重要性图
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

            features, importances_vals = zip(*top10)
            y_pos = np.arange(len(features))

            # Total Gain图
            bars1 = ax1.barh(y_pos, importances_vals, color='skyblue', alpha=0.8)
            ax1.set_yticks(y_pos)
            ax1.set_yticklabels(features)
            ax1.set_xlabel('Total Gain')
            ax1.set_title(f'{model_name} - Top 10 Features Total Gain')
            ax1.grid(True, alpha=0.3)
            ax1.invert_yaxis()

            # 添加数值标签
            for i, bar in enumerate(bars1):
                width = bar.get_width()
                ax1.text(width + width*0.01, bar.get_y() + bar.get_height()/2,
                        f'{width:.3f}', ha='left', va='center', fontsize=9)

            # Average Gain图
            avg_gains = [imp/n_features for imp in importances_vals]
            bars2 = ax2.barh(y_pos, avg_gains, color='lightgreen', alpha=0.8)
            ax2.set_yticks(y_pos)
            ax2.set_yticklabels(features)
            ax2.set_xlabel('Average Gain per Split')
            ax2.set_title(f'{model_name} - Top 10 Features Average Gain')
            ax2.grid(True, alpha=0.3)
            ax2.invert_yaxis()

            # 添加数值标签
            for i, bar in enumerate(bars2):
                width = bar.get_width()
                ax2.text(width + width*0.01, bar.get_y() + bar.get_height()/2,
                        f'{width:.4f}', ha='left', va='center', fontsize=9)

            plt.tight_layout()
            plt.show()

            break  # 只显示第一个支持特征重要性的模型

def find_best_model(results):
    """找出最佳模型"""

    print("\n" + "="*60)
    print("🏆 最佳模型推荐")
    print("="*60)

    # 基于验证集AUC-PR选择最佳模型
    best_score = 0
    best_model = None

    for model_name, model_results in results.items():
        val_score = model_results['Validation']['auc_pr']
        if val_score > best_score:
            best_score = val_score
            best_model = model_name

    if best_model:
        print(f"\n🥇 最佳模型: {best_model}")
        print("-" * 40)

        best_results = results[best_model]

        for dataset in ['Train', 'Validation', 'Test']:
            metrics = best_results[dataset]
            print(f"\n{dataset} Set:")
            print(f"  AUC-ROC: {metrics['auc_roc']:.4f}")
            print(f"  AUC-PR:  {metrics['auc_pr']:.4f}")
            print(f"  Acc:     {metrics['accuracy']:.4f}")
            print(f"  Sens:    {metrics['sensitivity']:.4f}")
            print(f"  Spec:    {metrics['specificity']:.4f}")

        # 稳定性检查
        train_auc = best_results['Train']['auc_pr']
        val_auc = best_results['Validation']['auc_pr']
        test_auc = best_results['Test']['auc_pr']

        print(f"\n📈 稳定性分析:")
        print(f"  Train-Val差异: {abs(train_auc - val_auc):.4f}")
        print(f"  Val-Test差异:  {abs(val_auc - test_auc):.4f}")

        if abs(train_auc - val_auc) > 0.1:
            print("  ⚠️ 可能过拟合")
        else:
            print("  ✅ 模型稳定")

    return best_model

# 主函数 - 一键运行所有分析
def run_complete_simple_analysis():
    """运行完整的简单分析"""

    print("🚀 开始完整的简单模型分析...")

    try:
        # 1. 创建新模型
        fresh_models = create_fresh_models()

        # 2. 准备数据
        X_train_scaled, X_val_scaled, X_test_scaled, feature_names = prepare_data_simple()

        # 3. 评估模型
        results = evaluate_simple_models(
            fresh_models, X_train_scaled, X_val_scaled, X_test_scaled,
            y_train, y_val, y_test
        )

        # 4. 打印汇总表
        print_summary_table(results)

        # 5. 绘制性能对比图
        print("\n📈 绘制性能对比图...")
        plot_performance_bars(results)

        # 6. 绘制混淆矩阵
        print("\n🔥 绘制混淆矩阵...")
        plot_confusion_matrices(results)

        # 7. 特征重要性分析
        analyze_feature_importance(fresh_models, feature_names)

        # 8. 最佳模型推荐
        best_model = find_best_model(results)

        print("\n✅ 完整分析成功完成!")

        return results, fresh_models, feature_names, best_model

    except Exception as e:
        print(f"❌ 分析失败: {str(e)}")
        import traceback
        traceback.print_exc()
        return None, None, None, None

# 立即执行分析
print("🎯 执行最简单的完整分析...")
final_results, final_models, final_features, final_best = run_complete_simple_analysis()
```

```python
# 基于A/T/N特征重要性的加权XGBoost模型
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, ConfusionMatrixDisplay
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
import warnings
warnings.filterwarnings('ignore')

def identify_atn_top_features():
    """根据A/T/N分析结果定义重要特征"""

    print("🔍 基于A/T/N分析结果的重要特征...")

    # 从你的图中提取的重要特征（请根据实际图片调整）
    atn_important_features = {
        # A类重要特征 (第一组图)
        'A_type_features': [
            'admission_vision',
            'age',
            'ast',
            'fibrinogen',
            'uric_acid'
        ],

        # T类重要特征 (第二组图)
        'T_type_features': [
            'admission_vision',
            'urine_specific_gravity',
            'spherical_equivalent',
            'neutrophil_abs',
            'large_platelet_ratio'
        ],

        # N类重要特征 (第三组图)
        'N_type_features': [
            'admission_vision',
            'spherical_equivalent',
            'age',
            'ast',
            'uric_acid'
        ]
    }

    # 统计各特征出现频次，作为权重依据
    feature_counts = {}
    all_features = []

    for category, features in atn_important_features.items():
        all_features.extend(features)
        print(f"{category}: {features}")

    for feature in set(all_features):
        feature_counts[feature] = all_features.count(feature)

    # 按出现频次排序，选择top特征
    top_atn_features = sorted(feature_counts.items(), key=lambda x: x[1], reverse=True)

    print(f"\n各特征在A/T/N分析中的重要性统计:")
    for feature, count in top_atn_features:
        print(f"  {feature}: 出现{count}次")

    return atn_important_features, feature_counts

def create_feature_weights(feature_names, atn_feature_counts, base_weight=1.0, boost_factor=5.0):
    """创建基于A/T/N重要性的特征权重"""

    print(f"\n🎯 创建特征权重 (基础权重: {base_weight}, 增强因子: {boost_factor})...")

    # 初始化所有特征为基础权重
    feature_weights = {feature: base_weight for feature in feature_names}

    # 根据A/T/N重要性提升权重
    boosted_features = []
    for feature, count in atn_feature_counts.items():
        if feature in feature_names:
            # 根据出现次数决定权重增强幅度
            weight_multiplier = base_weight + (boost_factor - base_weight) * (count / 3.0)
            feature_weights[feature] = weight_multiplier
            boosted_features.append((feature, weight_multiplier, count))

    print(f"增强权重的特征:")
    for feature, weight, count in sorted(boosted_features, key=lambda x: x[1], reverse=True):
        print(f"  {feature}: {weight:.2f} (出现{count}次)")

    return feature_weights

def apply_feature_weights_to_data(X, feature_weights):
    """将特征权重应用到数据上"""

    X_weighted = X.copy()

    if isinstance(X_weighted, pd.DataFrame):
        for feature, weight in feature_weights.items():
            if feature in X_weighted.columns:
                X_weighted[feature] = X_weighted[feature] * weight
    else:
        # 如果是numpy数组，需要特征名映射
        if hasattr(X, 'columns'):
            feature_names = X.columns
        else:
            print("警告: 无法获取特征名，跳过权重应用")
            return X

        for i, feature in enumerate(feature_names):
            if feature in feature_weights:
                X_weighted[:, i] = X_weighted[:, i] * feature_weights[feature]

    return X_weighted

def create_weighted_xgb_model(feature_weights, y_train):
    """创建应用特征权重的XGBoost模型"""

    print("🚀 创建加权XGBoost模型...")

    # 计算类别权重
    pos_ratio = y_train.mean()
    scale_pos_weight = (1 - pos_ratio) / pos_ratio

    # 创建增强的XGBoost模型
    weighted_xgb = XGBClassifier(
        objective='binary:logistic',
        eval_metric='auc',
        n_estimators=400,  # 增加树的数量
        learning_rate=0.03,  # 降低学习率
        max_depth=5,  # 增加深度
        min_child_weight=2,  # 降低最小权重
        subsample=0.85,
        colsample_bytree=0.85,
        scale_pos_weight=scale_pos_weight,
        reg_alpha=0.5,  # 降低正则化
        reg_lambda=0.5,
        random_state=42,
        use_label_encoder=False
    )

    print(f"模型参数调整:")
    print(f"  n_estimators: 400 (增加)")
    print(f"  learning_rate: 0.03 (降低)")
    print(f"  max_depth: 5 (增加)")
    print(f"  scale_pos_weight: {scale_pos_weight:.3f}")

    return weighted_xgb

def fine_tune_weighted_model(model, X_train_weighted, y_train, cv_folds=3):
    """对加权模型进行超参数微调"""

    print("🔧 开始超参数微调...")

    # 定义微调参数空间
    param_grid = {
        'n_estimators': [300, 400, 500],
        'learning_rate': [0.02, 0.03, 0.05],
        'max_depth': [4, 5, 6],
        'min_child_weight': [1, 2, 3],
        'subsample': [0.8, 0.85, 0.9],
        'colsample_bytree': [0.8, 0.85, 0.9],
        'reg_alpha': [0.3, 0.5, 0.8],
        'reg_lambda': [0.3, 0.5, 0.8]
    }

    # 使用随机搜索进行微调
    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_grid,
        n_iter=20,  # 减少搜索次数以节省时间
        cv=StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42),
        scoring='roc_auc',
        random_state=42,
        n_jobs=-1,
        verbose=1
    )

    # 执行搜索
    random_search.fit(X_train_weighted, y_train)

    print(f"最佳CV AUC: {random_search.best_score_:.4f}")
    print(f"最佳参数:")
    for param, value in random_search.best_params_.items():
        print(f"  {param}: {value}")

    return random_search.best_estimator_

def evaluate_weighted_model(model, X_train_weighted, X_val_weighted, X_test_weighted,
                          y_train, y_val, y_test, model_name="Weighted XGB"):
    """评估加权模型"""

    print(f"\n📊 评估{model_name}模型...")

    # 训练模型
    model.fit(X_train_weighted, y_train)

    # 评估三个数据集
    results = []
    datasets = [
        (X_train_weighted, y_train, "Train"),
        (X_val_weighted, y_val, "Valid"),
        (X_test_weighted, y_test, "Test")
    ]

    for X, y_true, tag in datasets:
        # 预测
        y_pred = model.predict(X)
        y_pred_proba = model.predict_proba(X)[:, 1]

        # 计算指标
        auc_score = roc_auc_score(y_true, y_pred_proba)
        cm = confusion_matrix(y_true, y_pred)
        tn, fp, fn, tp = cm.ravel()

        accuracy = (tp + tn) / (tp + tn + fp + fn)
        sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0

        results.append({
            "tag": tag,
            "auc": auc_score,
            "acc": accuracy,
            "sensitivity": sensitivity,
            "specificity": specificity,
            "precision": precision,
            "cm": cm
        })

        print(f"  {tag:5} - AUC: {auc_score:.4f}, Acc: {accuracy:.4f}, Sens: {sensitivity:.4f}")

    return results

def compare_models(original_results, weighted_results):
    """比较原始模型和加权模型的性能"""

    print("\n📈 模型性能对比:")
    print("="*70)

    # 创建对比表
    comparison_data = []

    for orig_res, weight_res in zip(original_results, weighted_results):
        dataset = orig_res['tag']
        comparison_data.append({
            'Dataset': dataset,
            'Original_AUC': orig_res['auc'],
            'Weighted_AUC': weight_res['auc'],
            'AUC_Improvement': weight_res['auc'] - orig_res['auc'],
            'Original_Acc': orig_res['acc'],
            'Weighted_Acc': weight_res['acc'],
            'Acc_Improvement': weight_res['acc'] - orig_res['acc']
        })

    comparison_df = pd.DataFrame(comparison_data)

    print("模型性能对比表:")
    print(comparison_df.round(4))

    # 可视化对比
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # AUC对比
    x = np.arange(len(comparison_df))
    width = 0.35

    ax1.bar(x - width/2, comparison_df['Original_AUC'], width, label='Original', alpha=0.8, color='skyblue')
    ax1.bar(x + width/2, comparison_df['Weighted_AUC'], width, label='Weighted', alpha=0.8, color='lightgreen')
    ax1.set_xlabel('Dataset')
    ax1.set_ylabel('AUC Score')
    ax1.set_title('AUC Score Comparison')
    ax1.set_xticks(x)
    ax1.set_xticklabels(comparison_df['Dataset'])
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # 添加数值标签
    for i, (orig, weight) in enumerate(zip(comparison_df['Original_AUC'], comparison_df['Weighted_AUC'])):
        ax1.text(i - width/2, orig + 0.01, f'{orig:.3f}', ha='center', va='bottom')
        ax1.text(i + width/2, weight + 0.01, f'{weight:.3f}', ha='center', va='bottom')

    # 准确率对比
    ax2.bar(x - width/2, comparison_df['Original_Acc'], width, label='Original', alpha=0.8, color='skyblue')
    ax2.bar(x + width/2, comparison_df['Weighted_Acc'], width, label='Weighted', alpha=0.8, color='lightgreen')
    ax2.set_xlabel('Dataset')
    ax2.set_ylabel('Accuracy Score')
    ax2.set_title('Accuracy Score Comparison')
    ax2.set_xticks(x)
    ax2.set_xticklabels(comparison_df['Dataset'])
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # 添加数值标签
    for i, (orig, weight) in enumerate(zip(comparison_df['Original_Acc'], comparison_df['Weighted_Acc'])):
        ax2.text(i - width/2, orig + 0.01, f'{orig:.3f}', ha='center', va='bottom')
        ax2.text(i + width/2, weight + 0.01, f'{weight:.3f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

    # 计算总体改善
    avg_auc_improvement = comparison_df['AUC_Improvement'].mean()
    avg_acc_improvement = comparison_df['Acc_Improvement'].mean()

    print(f"\n总体改善:")
    print(f"  平均AUC提升: {avg_auc_improvement:.4f}")
    print(f"  平均准确率提升: {avg_acc_improvement:.4f}")

    if avg_auc_improvement > 0:
        print("✅ 加权模型整体性能有所提升")
    else:
        print("⚠️ 加权模型未带来明显改善")

    return comparison_df

def run_weighted_model_analysis():
    """运行完整的加权模型分析"""

    print("🎯 开始基于A/T/N特征重要性的模型优化...")
    print("="*60)

    # 检查必要变量
    if not all(var in globals() for var in ['X_train', 'X_val', 'X_test', 'y_train', 'y_val', 'y_test']):
        print("❌ 缺少必要的数据变量")
        return None

    try:
        # 1. 识别A/T/N重要特征
        atn_features, feature_counts = identify_atn_top_features()

        # 2. 准备数据
        if hasattr(X_train, 'values'):
            feature_names = X_train.columns.tolist()
            X_train_np = X_train.values
            X_val_np = X_val.values
            X_test_np = X_test.values
        else:
            feature_names = [f'feature_{i}' for i in range(X_train.shape[1])]
            X_train_np = X_train
            X_val_np = X_val
            X_test_np = X_test

        # 3. 创建特征权重
        feature_weights = create_feature_weights(feature_names, feature_counts,
                                                base_weight=1.0, boost_factor=2.5)

        # 4. 应用特征权重
        print("\n⚖️ 应用特征权重...")
        X_train_weighted = apply_feature_weights_to_data(
            pd.DataFrame(X_train_np, columns=feature_names), feature_weights
        )
        X_val_weighted = apply_feature_weights_to_data(
            pd.DataFrame(X_val_np, columns=feature_names), feature_weights
        )
        X_test_weighted = apply_feature_weights_to_data(
            pd.DataFrame(X_test_np, columns=feature_names), feature_weights
        )

        # 标准化加权后的数据
        scaler = StandardScaler()
        X_train_weighted_scaled = scaler.fit_transform(X_train_weighted)
        X_val_weighted_scaled = scaler.transform(X_val_weighted)
        X_test_weighted_scaled = scaler.transform(X_test_weighted)

        # 5. 创建加权模型
        weighted_model = create_weighted_xgb_model(feature_weights, y_train)

        # 6. 微调模型
        print("\n🔧 开始模型微调...")
        tuned_weighted_model = fine_tune_weighted_model(
            weighted_model, X_train_weighted_scaled, y_train
        )

        # 7. 评估加权模型
        weighted_results = evaluate_weighted_model(
            tuned_weighted_model, X_train_weighted_scaled, X_val_weighted_scaled,
            X_test_weighted_scaled, y_train, y_val, y_test, "Tuned Weighted XGB"
        )

        # 8. 与原始模型比较
        if 'xgb_analysis' in globals() and xgb_analysis is not None:
            original_results = xgb_analysis['results']
            comparison_df = compare_models(original_results, weighted_results)
        else:
            print("⚠️ 未找到原始模型结果，无法进行比较")
            comparison_df = None

        print("\n✅ 加权模型分析完成!")

        return {
            'weighted_model': tuned_weighted_model,
            'weighted_results': weighted_results,
            'feature_weights': feature_weights,
            'atn_features': atn_features,
            'comparison': comparison_df,
            'scaler': scaler
        }

    except Exception as e:
        print(f"❌ 分析失败: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

# 立即执行加权模型分析
print("🚀 启动基于A/T/N特征重要性的模型优化...")
weighted_analysis = run_weighted_model_analysis()

if weighted_analysis:
    print("\n🎉 加权模型优化完成!")
    print("可用结果:")
    print("- weighted_analysis['weighted_model']: 优化后的模型")
    print("- weighted_analysis['weighted_results']: 性能结果")
    print("- weighted_analysis['feature_weights']: 特征权重")
    print("- weighted_analysis['comparison']: 性能对比")
```

```python
# 按指定格式绘制混淆矩阵和ROC曲线
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, roc_curve
import warnings
warnings.filterwarnings('ignore')

def prepare_results_for_plotting():
    """准备结果数据为指定格式"""

    # 检查是否有修复后的结果
    if 'fixed_weighted_display' in globals() and fixed_weighted_display is not None:
        results = fixed_weighted_display['results']
    elif 'weighted_analysis' in globals() and weighted_analysis is not None:
        # 如果没有修复的结果，使用原始结果但需要补充缺失的键
        print("使用原始结果并补充缺失数据...")

        # 重新生成包含所有必要键的结果
        weighted_model = weighted_analysis['weighted_model']
        feature_weights = weighted_analysis['feature_weights']

        # 准备数据
        if hasattr(X_train, 'values'):
            feature_names = X_train.columns.tolist()
            X_train_df = pd.DataFrame(X_train.values, columns=feature_names)
            X_val_df = pd.DataFrame(X_val.values, columns=feature_names)
            X_test_df = pd.DataFrame(X_test.values, columns=feature_names)
        else:
            feature_names = [f'feature_{i}' for i in range(X_train.shape[1])]
            X_train_df = pd.DataFrame(X_train, columns=feature_names)
            X_val_df = pd.DataFrame(X_val, columns=feature_names)
            X_test_df = pd.DataFrame(X_test, columns=feature_names)

        # 应用权重
        def apply_weights(X_df, weights):
            X_weighted = X_df.copy()
            for feature, weight in weights.items():
                if feature in X_weighted.columns:
                    X_weighted[feature] = X_weighted[feature] * weight
            return X_weighted

        from sklearn.preprocessing import StandardScaler
        from sklearn.metrics import roc_auc_score, confusion_matrix

        X_train_weighted = apply_weights(X_train_df, feature_weights)
        X_val_weighted = apply_weights(X_val_df, feature_weights)
        X_test_weighted = apply_weights(X_test_df, feature_weights)

        # 标准化
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train_weighted)
        X_val_scaled = scaler.transform(X_val_weighted)
        X_test_scaled = scaler.transform(X_test_weighted)

        # 训练并预测
        weighted_model.fit(X_train_scaled, y_train)

        results = []
        datasets = [
            (X_train_scaled, y_train, "Train"),
            (X_val_scaled, y_val, "Valid"),
            (X_test_scaled, y_test, "Test")
        ]

        for X, y_true, tag in datasets:
            y_pred = weighted_model.predict(X)
            y_pred_proba = weighted_model.predict_proba(X)[:, 1]

            auc_score = roc_auc_score(y_true, y_pred_proba)
            cm = confusion_matrix(y_true, y_pred)
            tn, fp, fn, tp = cm.ravel()
            accuracy = (tp + tn) / (tp + tn + fp + fn)

            results.append({
                "tag": tag,
                "true": y_true,
                "pred": y_pred,
                "prob": y_pred_proba,
                "auc": auc_score,
                "acc": accuracy,
                "cm": cm
            })
    else:
        print("未找到加权模型结果")
        return None, None, None

    # 按照指定变量名创建结果
    train_res = results[0]
    valid_res = results[1]
    test_res = results[2]

    return train_res, valid_res, test_res

def plot_exact_format():
    """按照用户指定的确切格式绘制图表"""

    print("按指定格式绘制混淆矩阵和ROC曲线...")

    # 准备数据
    train_res, valid_res, test_res = prepare_results_for_plotting()

    if train_res is None:
        print("无法准备数据")
        return

    print(f"Train AUC: {train_res['auc']:.3f}")
    print(f"Valid AUC: {valid_res['auc']:.3f}")
    print(f"Test AUC: {test_res['auc']:.3f}")

    # 1. 混淆矩阵 - 第一种格式
    print("\n=== 混淆矩阵 ===")

    for res in [train_res, valid_res, test_res]:
        ConfusionMatrixDisplay(res["cm"], display_labels=["0", "1"]).plot(cmap="Blues")
        plt.title(f'{res["tag"]} Confusion Matrix')
        plt.tight_layout()
        plt.show()

    # 2. ROC曲线 - 第一种格式
    print("\n=== ROC曲线 ===")

    plt.figure(figsize=(6, 5))
    for res, style in zip([train_res, valid_res, test_res], ["-", "--", "-."]):
        fpr, tpr, _ = roc_curve(res["true"], res["prob"])
        plt.plot(fpr, tpr, style, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
    plt.plot([0, 1], [0, 1], "k:", lw=1)
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC — XGBoost Bagging + Isotonic Calibration")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 3. 混淆矩阵 - 第二种格式（重复）
    print("\n=== 混淆矩阵（第二次） ===")

    for res in [train_res, valid_res, test_res]:
        ConfusionMatrixDisplay(res["cm"], display_labels=["0", "1"]).plot(cmap="Blues")
        plt.title(f'{res["tag"]} Confusion Matrix')
        plt.tight_layout()
        plt.show()

    # 4. ROC曲线 - 第二种格式（图例在右下角）
    print("\n=== ROC曲线（图例右下角） ===")

    plt.figure(figsize=(6, 5))
    for res, style in zip([train_res, valid_res, test_res], ["-", "--", "-."]):
        fpr, tpr, _ = roc_curve(res["true"], res["prob"])
        plt.plot(fpr, tpr, style, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
    plt.plot([0, 1], [0, 1], "k:", lw=1)
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC for N — XGBoost Bagging + Isotonic Calibration")
    plt.grid()
    plt.legend(loc="lower right")  # 图例放在右下角
    plt.tight_layout()
    plt.show()

    # 5. 性能汇总表
    print("\n=== 性能汇总 ===")

    summary_data = []
    for res in [train_res, valid_res, test_res]:
        tn, fp, fn, tp = res["cm"].ravel()
        accuracy = (tp + tn) / (tp + tn + fp + fn)
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0

        summary_data.append({
            "Dataset": res["tag"],
            "AUC": res["auc"],
            "Accuracy": accuracy,
            "Recall₁": recall
        })

    import pandas as pd
    summary = pd.DataFrame(summary_data).round(3)
    print("\n=== Performance Summary ===")
    print(summary.to_string(index=False))

    return train_res, valid_res, test_res, summary

# 立即执行绘制
print("开始按指定格式绘制图表...")
train_res, valid_res, test_res, summary = plot_exact_format()

if train_res is not None:
    print("\n图表绘制完成！")
    print("变量已创建:")
    print("- train_res: 训练集结果")
    print("- valid_res: 验证集结果")
    print("- test_res: 测试集结果")
    print("- summary: 性能汇总表")
else:
    print("绘制失败，请检查数据")
```

```python
# 简化版A/T/N特征排名检查
import pandas as pd
import numpy as np

def simple_atn_ranking_check():
    """简化版检查A/T/N特征在当前模型中的排名"""

    print("检查A/T/N重要特征在当前加权模型中的排名...")

    # 获取模型
    model = None
    if 'fixed_weighted_display' in globals() and fixed_weighted_display is not None:
        model = fixed_weighted_display['model']
    elif 'weighted_analysis' in globals() and weighted_analysis is not None:
        model = weighted_analysis['weighted_model']

    if model is None:
        print("未找到加权模型")
        return

    # 获取特征名称
    if hasattr(X_train, 'columns'):
        feature_names = X_train.columns.tolist()
    else:
        feature_names = [f'feature_{i}' for i in range(X_train.shape[1])]

    # 获取特征重要性
    importances = model.feature_importances_

    # 创建特征重要性排序
    feature_importance_data = []
    for i, (feature, importance) in enumerate(zip(feature_names[:len(importances)], importances)):
        feature_importance_data.append({
            'Feature': feature,
            'Importance': importance
        })

    # 按重要性排序
    feature_df = pd.DataFrame(feature_importance_data)
    feature_df = feature_df.sort_values('Importance', ascending=False).reset_index(drop=True)
    feature_df['Rank'] = range(1, len(feature_df) + 1)

    # 定义A/T/N重要特征
    atn_features = {
        'admission_vision': 3,  # 在A/T/N中出现3次
        'spherical_equivalent': 2,
        'age': 2,
        'ast': 2,
        'uric_acid': 2,
        'fibrinogen': 1,
        'urine_specific_gravity': 1,
        'neutrophil_abs': 1,
        'large_platelet_ratio': 1
    }

    print("\n" + "="*80)
    print("A/T/N重要特征在当前加权模型中的排名分析")
    print("="*80)
    print(f"{'特征名称':<25} {'A/T/N出现次数':<12} {'当前排名':<8} {'重要性得分':<12} {'状态'}")
    print("-" * 80)

    results = []

    for feature, atn_count in atn_features.items():
        # 查找该特征在当前排序中的位置
        feature_rows = feature_df[feature_df['Feature'] == feature]

        if len(feature_rows) > 0:
            rank = feature_rows.iloc[0]['Rank']
            importance = feature_rows.iloc[0]['Importance']

            # 判断状态
            if rank <= 10:
                status = "✓ Top10"
            elif rank <= 20:
                status = "○ Top20"
            else:
                status = "✗ 低排名"

            print(f"{feature:<25} {atn_count:<12} {rank:<8} {importance:<12.4f} {status}")

            results.append({
                'feature': feature,
                'atn_count': atn_count,
                'rank': rank,
                'importance': importance
            })
        else:
            print(f"{feature:<25} {atn_count:<12} {'未找到':<8} {'N/A':<12} ✗ 不存在")

    # 统计分析
    print("\n" + "="*50)
    print("统计分析")
    print("="*50)

    if results:
        # 计算统计
        ranks = [r['rank'] for r in results]
        atn_counts = [r['atn_count'] for r in results]

        avg_rank = np.mean(ranks)
        print(f"A/T/N特征平均排名: {avg_rank:.1f}")

        # 高频特征分析
        high_freq_results = [r for r in results if r['atn_count'] >= 2]
        if high_freq_results:
            high_freq_ranks = [r['rank'] for r in high_freq_results]
            avg_high_freq_rank = np.mean(high_freq_ranks)
            print(f"高频A/T/N特征(≥2次)平均排名: {avg_high_freq_rank:.1f}")

            print(f"\n高频A/T/N特征详情:")
            for r in high_freq_results:
                print(f"  {r['feature']}: 排名{r['rank']} (A/T/N出现{r['atn_count']}次)")

        # 排名分布
        top10_count = len([r for r in results if r['rank'] <= 10])
        top20_count = len([r for r in results if r['rank'] <= 20])

        print(f"\n排名分布:")
        print(f"  Top 10: {top10_count}/{len(results)} 个A/T/N特征")
        print(f"  Top 20: {top20_count}/{len(results)} 个A/T/N特征")

        # 权重效果分析
        print(f"\n权重效果分析:")

        # 检查最重要的A/T/N特征(admission_vision)的表现
        admission_vision_result = next((r for r in results if r['feature'] == 'admission_vision'), None)
        if admission_vision_result:
            rank = admission_vision_result['rank']
            print(f"  admission_vision (最重要A/T/N特征): 排名第{rank}位")
            if rank > 20:
                print("    ❌ 权重可能未生效 - 最重要特征排名很低")
            elif rank <= 10:
                print("    ✓ 权重可能生效 - 排名较高")
            else:
                print("    ○ 权重效果一般 - 排名中等")

        # 检查不在A/T/N中但排名很高的特征
        print(f"\n当前Top 10特征中非A/T/N特征:")
        top10_features = feature_df.head(10)['Feature'].tolist()
        non_atn_top10 = [f for f in top10_features if f not in atn_features.keys()]

        for i, feature in enumerate(non_atn_top10[:5]):  # 只显示前5个
            rank = i + 1
            for idx, row in feature_df.iterrows():
                if row['Feature'] == feature:
                    rank = row['Rank']
                    importance = row['Importance']
                    break
            print(f"  {rank}. {feature}: {importance:.4f}")

    return feature_df, results

# 执行简化检查
print("开始简化版A/T/N特征排名检查...")
feature_ranking_df, atn_results = simple_atn_ranking_check()

```

```python
# 保存加权模型所有分析结果
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from datetime import datetime
from sklearn.metrics import ConfusionMatrixDisplay, roc_curve, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

def create_weighted_save_directory():
    """创建加权模型结果保存目录"""
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    save_dir = f"weighted_xgb_analysis_results_{timestamp}"

    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    print(f"创建保存目录: {save_dir}")
    return save_dir

def prepare_weighted_results_for_saving():
    """准备加权模型结果数据"""

    print("准备加权模型数据...")

    # 获取模型
    model = None
    feature_weights = None

    if 'fixed_weighted_display' in globals() and fixed_weighted_display is not None:
        model = fixed_weighted_display['model']
    elif 'weighted_analysis' in globals() and weighted_analysis is not None:
        model = weighted_analysis['weighted_model']
        feature_weights = weighted_analysis.get('feature_weights', {})

    if model is None:
        print("未找到加权模型")
        return None, None, None, None

    # 获取特征名称
    if hasattr(X_train, 'columns'):
        feature_names = X_train.columns.tolist()
        X_train_df = pd.DataFrame(X_train.values, columns=feature_names)
        X_val_df = pd.DataFrame(X_val.values, columns=feature_names)
        X_test_df = pd.DataFrame(X_test.values, columns=feature_names)
    else:
        feature_names = [f'feature_{i}' for i in range(X_train.shape[1])]
        X_train_df = pd.DataFrame(X_train, columns=feature_names)
        X_val_df = pd.DataFrame(X_val, columns=feature_names)
        X_test_df = pd.DataFrame(X_test, columns=feature_names)

    # 应用特征权重（如果有的话）
    if feature_weights:
        print("应用特征权重...")
        for feature, weight in feature_weights.items():
            if feature in X_train_df.columns:
                X_train_df[feature] = X_train_df[feature] * weight
                X_val_df[feature] = X_val_df[feature] * weight
                X_test_df[feature] = X_test_df[feature] * weight

    # 标准化
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_df)
    X_val_scaled = scaler.transform(X_val_df)
    X_test_scaled = scaler.transform(X_test_df)

    # 训练模型并生成预测
    model.fit(X_train_scaled, y_train)

    # 生成结果
    results = []
    datasets = [
        (X_train_scaled, y_train, "Train"),
        (X_val_scaled, y_val, "Valid"),
        (X_test_scaled, y_test, "Test")
    ]

    for X, y_true, tag in datasets:
        y_pred = model.predict(X)
        y_pred_proba = model.predict_proba(X)[:, 1]

        auc_score = roc_auc_score(y_true, y_pred_proba)
        cm = confusion_matrix(y_true, y_pred)
        tn, fp, fn, tp = cm.ravel()
        accuracy = (tp + tn) / (tp + tn + fp + fn)

        results.append({
            "tag": tag,
            "true": y_true,
            "pred": y_pred,
            "prob": y_pred_proba,
            "auc": auc_score,
            "acc": accuracy,
            "cm": cm
        })

    return results, model, feature_names, feature_weights

def save_weighted_confusion_matrices(results, save_dir):
    """保存加权模型混淆矩阵"""

    print("保存混淆矩阵...")

    # 按用户要求的格式
    train_res, valid_res, test_res = results

    for res in [train_res, valid_res, test_res]:
        plt.figure(figsize=(6, 5))
        ConfusionMatrixDisplay(res["cm"], display_labels=["0", "1"]).plot(cmap="Blues")
        plt.title(f'{res["tag"]} Confusion Matrix')
        plt.tight_layout()

        filename = f"{save_dir}/confusion_matrix_{res['tag'].lower()}_weighted.png"
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"  保存: {filename}")
        plt.show()

def save_weighted_roc_curves(results, save_dir):
    """保存加权模型ROC曲线"""

    print("保存ROC曲线...")

    train_res, valid_res, test_res = results

    # 第一个ROC图
    plt.figure(figsize=(6, 5))
    for res, style in zip([train_res, valid_res, test_res], ["-", "--", "-."]):
        fpr, tpr, _ = roc_curve(res["true"], res["prob"])
        plt.plot(fpr, tpr, style, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
    plt.plot([0, 1], [0, 1], "k:", lw=1)
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC — Weighted XGBoost Model")
    plt.grid()
    plt.legend()
    plt.tight_layout()

    filename1 = f"{save_dir}/roc_curves_weighted_v1.png"
    plt.savefig(filename1, dpi=300, bbox_inches='tight')
    print(f"  保存: {filename1}")
    plt.show()

    # 第二个ROC图（图例在右下角）
    plt.figure(figsize=(6, 5))
    for res, style in zip([train_res, valid_res, test_res], ["-", "--", "-."]):
        fpr, tpr, _ = roc_curve(res["true"], res["prob"])
        plt.plot(fpr, tpr, style, label=f'{res["tag"]} AUC={res["auc"]:.2f}')
    plt.plot([0, 1], [0, 1], "k:", lw=1)
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC for Weighted XGBoost Model")
    plt.grid()
    plt.legend(loc="lower right")
    plt.tight_layout()

    filename2 = f"{save_dir}/roc_curves_weighted_v2.png"
    plt.savefig(filename2, dpi=300, bbox_inches='tight')
    print(f"  保存: {filename2}")
    plt.show()

def save_weighted_feature_importance(model, feature_names, save_dir):
    """保存加权模型特征重要性"""

    print("保存特征重要性图表...")

    importances = model.feature_importances_
    importance_df = pd.DataFrame({
        'Feature': feature_names[:len(importances)],
        'Gain': importances,
        'Total Gain': importances * 1000
    }).sort_values('Gain', ascending=False)

    # 保存特征重要性数据
    csv_filename = f"{save_dir}/feature_importance_weighted.csv"
    importance_df.to_csv(csv_filename, index=False)
    print(f"  保存数据: {csv_filename}")

    # 按用户要求的格式绘制图表
    top_k = 10
    top_features = importance_df.head(top_k)

    plt.figure(figsize=(12, 6))

    # --- Left: Gain ---
    plt.subplot(1, 2, 1)
    sns.barplot(data=top_features, x="Gain", y="Feature", palette="Blues_d")
    plt.title("Average Gain per Split")
    plt.xlabel("Gain")
    plt.ylabel("Feature")

    # --- Right: Total Gain ---
    plt.subplot(1, 2, 2)
    sns.barplot(data=top_features, x="Total Gain", y="Feature", palette="Greens_d")
    plt.title("Total Gain across All Splits")
    plt.xlabel("Total Gain")
    plt.ylabel("")

    plt.suptitle("Weighted XGBoost Feature Importance: Gain vs Total Gain", fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])

    feature_filename = f"{save_dir}/feature_importance_weighted_comparison.png"
    plt.savefig(feature_filename, dpi=300, bbox_inches='tight')
    print(f"  保存: {feature_filename}")
    plt.show()

    return importance_df

def save_atn_ranking_analysis(model, feature_names, save_dir):
    """保存A/T/N特征排名分析"""

    print("保存A/T/N特征排名分析...")

    # 获取特征重要性
    importances = model.feature_importances_
    feature_df = pd.DataFrame({
        'Feature': feature_names[:len(importances)],
        'Importance': importances
    }).sort_values('Importance', ascending=False).reset_index(drop=True)
    feature_df['Rank'] = range(1, len(feature_df) + 1)

    # A/T/N特征定义
    atn_features = {
        'admission_vision': 3,
        'spherical_equivalent': 2,
        'age': 2,
        'ast': 2,
        'uric_acid': 2,
        'fibrinogen': 1,
        'urine_specific_gravity': 1,
        'neutrophil_abs': 1,
        'large_platelet_ratio': 1
    }

    # 分析A/T/N特征排名
    atn_results = []
    for feature, atn_count in atn_features.items():
        feature_rows = feature_df[feature_df['Feature'] == feature]
        if len(feature_rows) > 0:
            rank = feature_rows.iloc[0]['Rank']
            importance = feature_rows.iloc[0]['Importance']
            atn_results.append({
                'Feature': feature,
                'ATN_Count': atn_count,
                'Current_Rank': rank,
                'Current_Importance': importance
            })

    # 保存A/T/N分析数据
    atn_df = pd.DataFrame(atn_results)
    atn_csv_filename = f"{save_dir}/atn_feature_ranking_analysis.csv"
    atn_df.to_csv(atn_csv_filename, index=False)
    print(f"  保存A/T/N数据: {atn_csv_filename}")

    # 创建A/T/N排名可视化
    if len(atn_results) > 0:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # 排名图
        colors = ['red' if x >= 2 else 'orange' for x in atn_df['ATN_Count']]
        bars1 = ax1.bar(range(len(atn_df)), atn_df['Current_Rank'], color=colors, alpha=0.7)
        ax1.set_xticks(range(len(atn_df)))
        ax1.set_xticklabels(atn_df['Feature'], rotation=45, ha='right')
        ax1.set_ylabel('当前排名 (数字越小越重要)')
        ax1.set_title('A/T/N特征在加权模型中的排名')
        ax1.axhline(y=15, color='green', linestyle='--', alpha=0.7, label='期望排名线(Top 15)')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # 重要性得分图
        bars2 = ax2.bar(range(len(atn_df)), atn_df['Current_Importance'], color=colors, alpha=0.7)
        ax2.set_xticks(range(len(atn_df)))
        ax2.set_xticklabels(atn_df['Feature'], rotation=45, ha='right')
        ax2.set_ylabel('特征重要性得分')
        ax2.set_title('A/T/N特征的重要性得分')
        ax2.grid(True, alpha=0.3)

        plt.tight_layout()

        atn_viz_filename = f"{save_dir}/atn_feature_ranking_visualization.png"
        plt.savefig(atn_viz_filename, dpi=300, bbox_inches='tight')
        print(f"  保存: {atn_viz_filename}")
        plt.show()

    return atn_df

def save_performance_summary(results, save_dir):
    """保存性能汇总表"""

    print("保存性能汇总表...")

    train_res, valid_res, test_res = results

    # 创建性能汇总
    summary_data = []
    for res in [train_res, valid_res, test_res]:
        tn, fp, fn, tp = res["cm"].ravel()
        accuracy = (tp + tn) / (tp + tn + fp + fn)
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        sensitivity = recall
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        summary_data.append({
            "Dataset": res["tag"],
            "AUC": res["auc"],
            "Accuracy": accuracy,
            "Recall₁": recall,
            "Sensitivity": sensitivity,
            "Specificity": specificity,
            "Precision": precision,
            "F1_Score": f1_score,
            "Samples": len(res["true"])
        })

    summary_df = pd.DataFrame(summary_data).round(4)

    # 保存CSV
    summary_csv = f"{save_dir}/performance_summary_weighted.csv"
    summary_df.to_csv(summary_csv, index=False)
    print(f"  保存: {summary_csv}")

    # 保存详细文本报告
    report_txt = f"{save_dir}/performance_detailed_report_weighted.txt"
    with open(report_txt, 'w', encoding='utf-8') as f:
        f.write("=== 加权XGBoost模型性能详细报告 ===\n\n")
        f.write(f"生成时间: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n")

        f.write("性能汇总:\n")
        f.write(summary_df.to_string(index=False))
        f.write("\n\n")

        # 混淆矩阵详情
        for res in [train_res, valid_res, test_res]:
            f.write(f"{res['tag']} 数据集混淆矩阵:\n")
            tn, fp, fn, tp = res["cm"].ravel()
            f.write(f"  真阴性 (TN): {tn}\n")
            f.write(f"  假阳性 (FP): {fp}\n")
            f.write(f"  假阴性 (FN): {fn}\n")
            f.write(f"  真阳性 (TP): {tp}\n\n")

    print(f"  保存: {report_txt}")

    return summary_df

def create_weighted_analysis_report(save_dir, summary_df, importance_df, atn_df, feature_weights):
    """创建完整的加权模型分析报告"""

    print("创建完整分析报告...")

    report_filename = f"{save_dir}/Weighted_XGBoost_Analysis_Report.md"

    with open(report_filename, 'w', encoding='utf-8') as f:
        f.write("# 加权XGBoost疾病诊断模型分析报告\n\n")
        f.write(f"**生成时间**: {datetime.now().strftime('%Y年%m月%d日 %H:%M:%S')}\n\n")

        f.write("## 1. 模型概述\n\n")
        f.write("本模型基于A/T/N特征重要性分析结果，对特定特征进行加权优化的XGBoost分类模型。\n\n")

        if feature_weights:
            f.write("### 应用的特征权重\n\n")
            f.write("| 特征名称 | 权重系数 | A/T/N出现频次 |\n")
            f.write("|----------|----------|---------------|\n")
            for feature, weight in sorted(feature_weights.items(), key=lambda x: x[1], reverse=True):
                f.write(f"| {feature} | {weight:.2f} | - |\n")
            f.write("\n")

        f.write("## 2. 模型性能\n\n")
        f.write("| 数据集 | AUC | 准确率 | 敏感性 | 特异性 |\n")
        f.write("|--------|-----|--------|--------|--------|\n")
        for _, row in summary_df.iterrows():
            f.write(f"| {row['Dataset']} | {row['AUC']:.4f} | {row['Accuracy']:.4f} | {row['Sensitivity']:.4f} | {row['Specificity']:.4f} |\n")

        f.write("\n## 3. 性能评估\n\n")
        test_auc = summary_df[summary_df['Dataset'] == 'Test']['AUC'].iloc[0]
        train_auc = summary_df[summary_df['Dataset'] == 'Train']['AUC'].iloc[0]

        if test_auc > 0.8:
            f.write("✅ **模型性能优秀**: 测试集AUC > 0.8\n\n")
        elif test_auc > 0.7:
            f.write("✅ **模型性能良好**: 测试集AUC > 0.7\n\n")
        elif test_auc > 0.6:
            f.write("⚠️ **模型性能中等**: 测试集AUC在0.6-0.7之间\n\n")
        else:
            f.write("❌ **模型性能较差**: 测试集AUC < 0.6\n\n")

        # 过拟合检查
        if abs(train_auc - test_auc) > 0.1:
            f.write("⚠️ **过拟合风险**: 训练集和测试集AUC差异较大\n\n")
        else:
            f.write("✅ **模型稳定**: 训练集和测试集性能相近\n\n")

        f.write("## 4. 特征重要性 (Top 10)\n\n")
        f.write("| 排名 | 特征名称 | 重要性得分 |\n")
        f.write("|------|----------|------------|\n")
        top10_features = importance_df.head(10)
        for i, (_, row) in enumerate(top10_features.iterrows(), 1):
            f.write(f"| {i} | {row['Feature']} | {row['Gain']:.4f} |\n")

        if len(atn_df) > 0:
            f.write("\n## 5. A/T/N特征分析\n\n")
            f.write("| A/T/N特征 | 当前排名 | 重要性得分 | A/T/N出现频次 |\n")
            f.write("|-----------|----------|------------|---------------|\n")
            for _, row in atn_df.iterrows():
                f.write(f"| {row['Feature']} | {row['Current_Rank']} | {row['Current_Importance']:.4f} | {row['ATN_Count']} |\n")

            # 权重效果分析
            f.write("\n### 权重应用效果\n\n")
            high_freq_low_rank = atn_df[(atn_df['ATN_Count'] >= 2) & (atn_df['Current_Rank'] > 20)]
            if len(high_freq_low_rank) > 0:
                f.write("⚠️ **发现问题**: 以下高频A/T/N特征排名较低，权重可能未充分生效:\n\n")
                for _, row in high_freq_low_rank.iterrows():
                    f.write(f"- {row['Feature']}: A/T/N出现{row['ATN_Count']}次，当前排名第{row['Current_Rank']}位\n")
                f.write("\n")

        f.write("## 6. 文件说明\n\n")
        f.write("- `confusion_matrix_*_weighted.png`: 各数据集混淆矩阵\n")
        f.write("- `roc_curves_weighted_*.png`: ROC曲线图\n")
        f.write("- `feature_importance_weighted_*.png`: 特征重要性图\n")
        f.write("- `atn_feature_ranking_*.png/csv`: A/T/N特征排名分析\n")
        f.write("- `performance_summary_weighted.csv`: 性能指标数据\n")
        f.write("- `performance_detailed_report_weighted.txt`: 详细性能报告\n")

    print(f"  保存完整报告: {report_filename}")

def save_all_weighted_results():
    """保存所有加权模型分析结果的主函数"""

    print("开始保存所有加权XGBoost模型分析结果...")
    print("=" * 60)

    # 创建保存目录
    save_dir = create_weighted_save_directory()

    try:
        # 准备数据
        results, model, feature_names, feature_weights = prepare_weighted_results_for_saving()

        if results is None:
            print("无法准备数据")
            return None

        print(f"\n模型性能快览:")
        for res in results:
            print(f"  {res['tag']}: AUC={res['auc']:.4f}, Accuracy={res['acc']:.4f}")

        # 1. 保存混淆矩阵
        save_weighted_confusion_matrices(results, save_dir)

        # 2. 保存ROC曲线
        save_weighted_roc_curves(results, save_dir)

        # 3. 保存特征重要性
        importance_df = save_weighted_feature_importance(model, feature_names, save_dir)

        # 4. 保存A/T/N排名分析
        atn_df = save_atn_ranking_analysis(model, feature_names, save_dir)

        # 5. 保存性能汇总
        summary_df = save_performance_summary(results, save_dir)

        # 6. 创建完整报告
        create_weighted_analysis_report(save_dir, summary_df, importance_df, atn_df, feature_weights)

        print(f"\n✅ 所有加权模型结果已保存到目录: {save_dir}")

        # 列出所有文件
        import glob
        files = glob.glob(f"{save_dir}/*")
        print(f"\n📂 保存的文件 ({len(files)}个):")
        for file in sorted(files):
            print(f"   - {os.path.basename(file)}")

        print(f"\n📄 查看完整报告: {save_dir}/Weighted_XGBoost_Analysis_Report.md")

        return save_dir

    except Exception as e:
        print(f"❌ 保存过程中出错: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

# 立即执行保存
print("开始保存加权XGBoost模型的所有分析结果...")
weighted_save_directory = save_all_weighted_results()

if weighted_save_directory:
    print(f"\n🎉 保存完成！所有文件都在 '{weighted_save_directory}' 目录中")
    print("包含:")
    print("- 混淆矩阵图 (按用户指定格式)")
    print("- ROC曲线图 (两个版本)")
    print("- 特征重要性对比图 (Gain vs Total Gain)")
    print("- A/T/N特征排名分析")
    print("- 性能汇总表和详细报告")
    print("- 完整的Markdown分析报告")
else:
    print("保存失败，请检查数据和模型")
```
